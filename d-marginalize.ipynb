```python
import scipy.linalg
import scipy.stats
import emcee
import triangle
```

```python
xs = np.array([-1,0,1])
err = 0.1
ftrue = 3
atrue = 0.
btrue = 0.
def lin_model(f):
    return np.array([[1,1,1],
                     [1+f**2,0,1+f**3]]).T
def model(f,a,b):
    return np.dot(lin_model(f),[a,b])
ys = model(ftrue,atrue,btrue)+np.random.randn(len(xs))*err
plt.plot(xs,ys,".")
```

```python
def logp(x):
    f,a,b = x
    ty = model(f,a,b)
    v = -0.5*np.sum(np.abs((ty-ys))**2)/err**2
    if np.isnan(v):
        return -np.inf
    else:
        return v
```

```python
ndim = 3
nwalkers = 20
p0 = [1e-6*np.random.randn(ndim)+np.array([ftrue,atrue,btrue]) 
      for i in range(nwalkers)]

sampler = emcee.EnsembleSampler(nwalkers, ndim, logp)
pos, prob, state = sampler.run_mcmc(p0, 10000)
sampler.reset()
sampler.run_mcmc(pos, 100000);
```

```python
def logp2(x):
    f, = x
    if not np.isfinite(f**4):
        return -np.inf
    A = lin_model(f)
    if not np.all(np.isfinite(A)):
        return -np.inf
    X, _, _, _ = scipy.linalg.lstsq(A,ys)
    ty = np.dot(A,X)
    v = -0.5*np.sum((ty-ys)**2)/err**2
    s, d = np.linalg.slogdet(np.dot(A.T, A))
    v -= 0.5*d
    if np.isnan(v):
        return -np.inf
    else:
        return v
```

```python
p0 = [1e-6*np.random.randn(1)+np.array([ftrue]) for i in range(nwalkers)]

sampler2 = emcee.EnsembleSampler(nwalkers, 1, logp2)

pos, prob, state = sampler2.run_mcmc(p0, 10000)
sampler2.reset()
sampler2.run_mcmc(pos, 100000);
```

# Analytical marginalization

Suppose I have a nonlinear least-squares problem depending on parameter vectors $x$ and $y$ where the values take the form $v(x,y) = f(x)+M(x)(y-y_0(x))$, where the goal value is $v_0$, $M(x)$ is a matrix, and $f(x)$ and $y_0(x)$ is chosen that for any $x$ and $y$, $||v(x,y)-v_0||\geq ||v(x,y_0(x))-v_0||$; such values can be obtained by applying linear least-squares minimization as part of the computation of $f(x)$. This condition implies that $||v(x,y)-v_0||^2 = ||f(x)-v_0||^2 + ||M(x)(y-y_0(x)||^2$. 

Now, we want to draw $(x,y)$ from a distribution whose PDF is proportional to $p(x,y) = e^{-||v(x,y)-v_0||^2/2}$ and then discard the $y$ values. We can write $p(x,y) = e^{-||f(x)-v_0||^2/2}e^{-||M(x)(y-y_0)||^2/2}$.

By those norm brackets, I mean an error distance; in my case, that's just a matter of dividing each component by the uncertainty on that part of the residual; I'm going to represent that by a matrix $\Sigma^{-1}$ so that $||w||^2 = w^T\Sigma^{-1}w$. 

We can therefore sample from $p(x) = \int p(x,y) dy = e^{-||f(x)-v_0||^2/2}L(x)$ where $L(x)=\int e^{-||M(x)(y-y_0)||^2/2} dy = \sqrt{(2\pi)^k \det(M(x)^T\Sigma^{-1}M(x))^{-1}}$.
