```python
duration = 60
fps = 30
show_obs = True
slow_time = 20 #s
slow_orbits = 2

fg_color = "white"
bg_color = "black"
psr_color = fg_color
c1_color = "yellow"
c2_color = "red"
cm_color = "gray"
ao_color = "cornflowerblue"
gbt_color = "limegreen"
wsrt_color = "orchid"

start_time = F.mjds[0]-F.base_mjd
transition_time = F.mjds[0]-F.base_mjd + F.best_parameters['pb_i']*slow_orbits
end_time = F.mjds[-1]-F.base_mjd

obs_markersize = 5
au = 499.004784
rsun = 2.32274689
fig = plt.figure()
ax_big = fig.add_axes([0,0,3./4.,1])
ax_big.set_axis_bgcolor(bg_color)

c2_trail, = ax_big.plot([],[],color=c2_color)
cm_trail, = ax_big.plot([],[],color=cm_color)
psr, = ax_big.plot([],[],'o',color=psr_color)
c1, = ax_big.plot([],[],'o',color=c1_color)
c2, = ax_big.plot([],[],'o',color=c2_color,zorder=50000)

ax_big.plot([-550,-550],[-au/2,au/2],color=fg_color,linewidth=2)
ax_big.text(-550,1.05*au/2,"1 AU", 
    verticalalignment='bottom', horizontalalignment='center', color=fg_color, fontsize=16)
ax_big.set_xlim(-600,600)
ax_big.set_ylim(-600,600)
ax_big.set_xticklabels([])
ax_big.set_yticklabels([])

ax_zoom = fig.add_axes([3./4.,2./3,1./4,1./3])
ax_zoom.set_axis_bgcolor(bg_color)
psr_zoom_trail, = ax_zoom.plot([],[],color=psr_color)
c1_zoom_trail, = ax_zoom.plot([],[],color=c1_color)
cm_zoom_trail, = ax_zoom.plot([],[],color=cm_color)
psr_zoom, = ax_zoom.plot([],[],'o',color=psr_color)
c1_zoom, = ax_zoom.plot([],[],'o',color=c1_color,zorder=50000)
ax_zoom.set_xlim(-20,20)
ax_zoom.set_ylim(-20,20)
ax_zoom.set_xticklabels([])
ax_zoom.set_yticklabels([])
ax_zoom.plot([16,16],[5*rsun-rsun/2,5*rsun+rsun/2],color=fg_color,linewidth=2)
ax_zoom.text(16,5*rsun+rsun/2,r"$R_\odot$", 
    verticalalignment='bottom', horizontalalignment='center', color=fg_color, fontsize=16) 
cm = (rotated_states[:,6,None]*rotated_states[:,:3]+
        rotated_states[:,13,None]*rotated_states[:,7:10])/(rotated_states[:,6,None]+
                                                            rotated_states[:,13,None])
cm_normal = (rotated_states_normal[:,6,None]*rotated_states_normal[:,:3]+
        rotated_states_normal[:,13,None]*rotated_states_normal[:,7:10])/(rotated_states_normal[:,6,None]+
                                                            rotated_states_normal[:,13,None])
first_new_obs = 0

ax_text = fig.add_axes([3./4.,0,1./4,2./3])
ax_text.set_axis_bgcolor(bg_color)
ax_text.set_xticks([])
ax_text.set_yticks([])
mjdreadout = ax_text.text(0, 1-0.01, "",
                      horizontalalignment="left",
                      verticalalignment="top",
                      fontsize=18,
                      color=fg_color,
                      transform=ax_text.transAxes)
extratext = ax_text.text(0, 0.75, r"""$P_o$ = 327 day
$P_i$ = 1.6 day
""",
                      horizontalalignment="left",
                      verticalalignment="top",
                      fontsize=16,
                      color=fg_color,
                      transform=ax_text.transAxes)

mp_text = ax_text.text(0, 0.60, r"$m_p$ = 1.443 $M_\odot$",
                      horizontalalignment="left",
                      verticalalignment="top",
                      fontsize=16,
                      color=psr_color,
                      transform=ax_text.transAxes)
m1_text = ax_text.text(0, 0.53, r"$m_1$ = 0.198 $M_\odot$",
                      horizontalalignment="left",
                      verticalalignment="top",
                      fontsize=16,
                      color=c1_color,
                      transform=ax_text.transAxes)
m2_text = ax_text.text(0, 0.46, r"$m_2$ = 0.411 $M_\odot$",
                      horizontalalignment="left",
                      verticalalignment="top",
                      fontsize=16,
                      color=c2_color,
                      transform=ax_text.transAxes)

credit = ax_text.text(0.99,0.01,"video by\nAnne Archibald",
                      horizontalalignment="right",
                      verticalalignment="bottom",
                      fontsize=14,
                      color=fg_color,
                      transform=ax_text.transAxes)
#speed = ax_text.text(1./2,0.15,"",
#                      horizontalalignment="center",
#                      verticalalignment="top",
#                      fontsize=18,
#                      color = fg_color,
#                      transform=ax_text.transAxes)

if show_obs:
    ao_text = ax_text.text(1./6,0.25,"AO",
                      horizontalalignment="center",
                      verticalalignment="top",
                      fontsize=14,
                      color = ao_color,
                      transform=ax_text.transAxes)
    gbt_text = ax_text.text(3./6,0.25,"GBT",
                      horizontalalignment="center",
                      verticalalignment="top",
                      fontsize=14,
                      color = gbt_color,
                      transform=ax_text.transAxes)
    wsrt_text = ax_text.text(5./6,0.25,"WSRT",
                      horizontalalignment="center",
                      verticalalignment="top",
                      fontsize=14,
                      color = wsrt_color,
                      transform=ax_text.transAxes)

def init():
    pass
def update(n):
    if n<slow_time*fps:
        time = start_time+(transition_time-start_time)*float(n)/(slow_time*fps)
        #speed.set_text(u"\u25B8")
    else:
        time = transition_time+(end_time-transition_time)*float(n-slow_time*fps)/((duration-slow_time)*fps)
        #speed.set_text(u"\u23E9")
    i0 = np.searchsorted(bbats,start_time)
    i = np.searchsorted(bbats,time)
    i = min(i,len(bbats)-1)
    
    psr.set_data(rotated_states[i,0],rotated_states[i,1])
    c1.set_data(rotated_states[i,7],rotated_states[i,8])
    c2.set_data(rotated_states[i,14],rotated_states[i,15])
    c2_trail.set_data(rotated_states[i0:i,14],rotated_states[i0:i,15])
    cm_trail.set_data(cm[i0:i,0],cm[i0:i,1])
    
    psr_zoom.set_data(rotated_states[i,0]-cm[i,0],rotated_states[i,1]-cm[i,1])
    c1_zoom.set_data(rotated_states[i,7]-cm[i,0],rotated_states[i,8]-cm[i,1])
    psr_zoom_trail.set_data(rotated_states[i0:i,0]-cm[i0:i,0],rotated_states[i0:i,1]-cm[i0:i,1])
    c1_zoom_trail.set_data(rotated_states[i0:i,7]-cm[i0:i,0],rotated_states[i0:i,8]-cm[i0:i,1])
    cm_zoom_trail.set_data(cm[i0:i,0]-cm[i,0],cm[i0:i,1]-cm[i,1])

    global first_new_obs
    while show_obs and first_new_obs<len(F.mjds) and F.mjds[first_new_obs]-F.base_mjd<time:
        t = F.tel_list[F.tels[first_new_obs]]
        if t.startswith("AO"):
            col = ao_color
        elif t.startswith("GBT"):
            col = gbt_color
        elif t.startswith("WSRT"):
            col = wsrt_color
        ax_big.plot([rotated_states_normal[first_new_obs,14]],
                    [rotated_states_normal[first_new_obs,15]],
                    '.',
                    color=col,
                    markersize=obs_markersize,zorder=first_new_obs+100)
        ax_zoom.plot([rotated_states_normal[first_new_obs,7]-cm_normal[first_new_obs,0]],
                     [rotated_states_normal[first_new_obs,8]-cm_normal[first_new_obs,1]],
                     '.',
                     color=col,
                     markersize=obs_markersize,zorder=first_new_obs+100)
        first_new_obs += 1
    mjdreadout.set_text("""PSR J0337+17
MJD %5.1f
%d TOAs
""" % (bbats[i]+F.base_mjd,np.sum(F.mjds-F.base_mjd<=bbats[i])))

w = 8
fig.set_size_inches(w,w*3./4)

from matplotlib import animation

anim = animation.FuncAnimation(fig, update, init_func=init, frames=int(duration*fps), blit=True)
anim.save('animation.webm', fps=fps, clear_temp=False, frame_prefix="animation/frame") #, extra_args=['-vcodec', 'libx264'])
#update(900)

plt.show()
```
