```python
from glob import glob
import os

import scipy.linalg

import emcee

import kepler
import threebody
```

```python
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 120
```

```python
def load(jobid):
    offsets, probs = [], []
    b = os.path.join("emcee-chains",jobid+".*")
    for o,p in zip(sorted(glob(os.path.join(b,"*-pos.npy"))),
                   sorted(glob(os.path.join(b,"*-prob.npy")))):
        offsets.append(np.load(o))
        probs.append(np.load(p))
    return np.array(offsets), np.array(probs)
```

```python
F = threebody.Fitter()
```

```python
for (i,p) in enumerate(F.parameters):
    e = F.best_errors[p]
    plt.subplot(131)
    plt.title("mean")
    plt.plot(np.mean(offsets[...,i]/e,axis=1))
    plt.subplot(132)
    plt.title("std")
    plt.semilogy(np.std(offsets[...,i]/e,axis=1))
eigvals = []
for j in range(offsets.shape[0]):
    s = np.array([F.best_errors[p]**(-1) for p in F.parameters])
    vals, vecs = scipy.linalg.eigh(np.cov(offsets[j].T)*s[:,None]*s[None,:])
    eigvals.append(vals)
eigvals = np.array(eigvals)
for v in eigvals.T:
    plt.subplot(133)
    plt.title("eigenvalues of cov. mtx.")
    plt.semilogy(v)
plt.gcf().set_size_inches(8,4)
plt.savefig("progress.pdf")
```

```python
offs = offsets[-50:].reshape((-1,offsets.shape[-1]))
stds = np.std(offs,axis=0)
vals, vecs = scipy.linalg.eigh(np.corrcoef(offs.T))
print vals[-1]
sv = vecs[:,-1]*stds
for (v,p) in sorted(zip(sv,F.parameters)):
    print p,v/F.best_errors[p]
```

```python
params = F.best_parameters.copy()
for (i,p) in enumerate(F.parameters):
    params[p] += vecs[i,-1]*stds[i]
r = F.residuals(params)
r0 = F.residuals(F.best_parameters)
```

```python
plt.hexbin(
    ((F.mjds-F.best_parameters['tasc_i'])/F.best_parameters['pb_i'])%1, 
    F.mjds,
    C=(r-r0), gridsize=20)
plt.colorbar()
print np.sum((r0/F.phase_uncerts)**2), np.sum((r/F.phase_uncerts)**2)
plt.savefig("change.pdf")
```
