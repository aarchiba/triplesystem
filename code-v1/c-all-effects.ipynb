```python
from glob import glob
import os
import pickle
from cStringIO import StringIO

import scipy.linalg

import emcee

import kepler
import threebody

import minuit_wrap
```

```python
job_dir = "emcee-chains/12504.nimrod" 
```

```python
with open(os.path.join(job_dir,"pulsars.pickle"), "rb") as f:
    pulsars = pickle.load(f)
```

```python
with open(os.path.join(job_dir,"fitter_params.pickle")) as f:
    fitter_params = pickle.load(f)
F = threebody.Fitter(**fitter_params)
```

```python
def display_parameter(p):
    if p not in pulsars:
        raise ValueError("Unknown parameter")
    best_i = np.argmax(pulsars["prob"])
    bp = F.best_parameters.copy()
    for k in bp:
        if k in pulsars:
            bp[k] = pulsars[k][best_i]
    best_resids = F.residuals(bp)
    
    m, sd = np.mean(pulsars[p]), np.std(pulsars[p])
    c = np.abs(pulsars[p]-m) >= sd
    best_j = np.argmax(pulsars["prob"][c])
    bp2 = F.best_parameters.copy()
    for k in bp2:
        if k in pulsars:
            bp2[k] = pulsars[k][c][best_j]
    offset_resids = F.residuals(bp2)
    
    
    stds = {}
    means = {}
    zs = {}
    for p2 in pulsars:
        stds[p2] = np.std(pulsars[p2])
        means[p2] = np.mean(pulsars[p2])
        zs[p2] = (pulsars[p2][c][best_j]-means[p2])/stds[p2]
    plt.hexbin(
        ((F.mjds-F.base_mjd-F.best_parameters['tasc_i'])/
         F.best_parameters['pb_i'])%1, 
        F.mjds,
        C=(offset_resids-best_resids)/F.best_parameters['f0']*1e6, 
        gridsize=20)
    plt.title("Change in residuals due to %.2g-sigma change in %s"
              %(zs[p],p))
    plt.colorbar().set_label(r"$\mu$s")
    plt.xlabel("Inner orbital phase")
    plt.ylabel("MJD")
    plt.cool()
    
    summary = StringIO()
    for z, p2 in sorted([(z,p2) for (p2,z) in zs.items()]):
        summary.write("%s %f\n" % (p2,z))
    #print summary.getvalue()
```

```python
display_parameter('delta')
```

```python
for k in pulsars:
    if k=='prob':
        continue
    plt.figure()
    display_parameter(k)
    plt.savefig("stat-change-%s.pdf" % k)
```
