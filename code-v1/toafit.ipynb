```python
def scale(p):
    return (np.array(p,np.float128)-np.array(best_parameters,np.float128))/np.array(threebody.best_errors,np.float128)
def unscale(p):
    return np.array(p,np.float128)*np.array(threebody.best_errors,np.float128)+np.array(best_parameters,np.float128)
```

```python
jump_vec = tels[:,None]==np.arange(1,len(tel_list))[None,:]
def compute_residuals(p,l,special=True, general=True, use_t_psr=True):
    (tzrmjd, f0, f1, 
            j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400) = l
    fix_bbats = bbats-np.dot(jump_vec,[j_AO327,j_GBT1500,j_GBT350,j_GBT820,j_WSRT1400])/86400.
    o = threebody.compute_orbit_bbat(p,fix_bbats,special=special,general=general)
    if use_t_psr:
        t_psr_s = (o["t_psr"]-tzrmjd)*86400
    else:
        t_psr_s = (o["t_d"]+o["states"][:,21]-tzrmjd)*86400
    phase = f0*t_psr_s+0.5*f1*t_psr_s**2
    return phase - pulses
    
```

```python
def mfun(asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan,
            tzrmjd, f0, f1, 
            j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400):
    p = unscale((asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan))
    (tzrmjd, f0, f1, 
            j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400) = unscale_lin((tzrmjd, f0, f1, 
                j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400))
    resids = compute_residuals(p,(tzrmjd, f0, f1, 
            j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400))
    r = np.sum((resids/p_uncerts)**2)
    global best_found_val, best_found_params, best_found_lin_params
    if r<best_found_val:
        best_found_val = r
        best_found_params = p
        best_found_lin_params = (tzrmjd, f0, f1, 
            j_AO327, j_GBT1500, j_GBT350, j_GBT820, j_WSRT1400)
    return r
```
