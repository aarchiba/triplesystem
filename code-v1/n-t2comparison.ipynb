```python
import subprocess
import os

import residuals
```

```python
timfile = "0337+17-scott-2013-06-06.tim"
parfile = "0337_bogus.par"
```

```python
subprocess.check_call(["tempo", "-f", parfile, timfile])
r = residuals.read_residuals()
t_bats = np.asarray(r.bary_TOA,np.float128)
print t_bats.shape
```

```python
def mjd_fromstring(s):
    i,f = s.split(".")
    i = int(i)
    f = np.float128("0."+f)
    return i+f
```

```python
os.environ['TEMPO2'] = '/home/aarchiba/software/tempo2/t2dir'
o = subprocess.check_output(["/home/aarchiba/software/tempo2/tempo2/tempo2",
    "-output", "general2", "-s", "OUTPUT {bat} {freq} {err}\n",
    "-tempo1",
    "-f", parfile, timfile])
t2_bats = []
freqs = []
errs = []
for l in o.split("\n"):
    if not l.startswith("OUTPUT"):
        continue
    t2_bats.append(mjd_fromstring(l.split()[1]))
    freqs.append(float(l.split()[2]))
    errs.append(float(l.split()[3]))
t2_bats = np.array(t2_bats)
print t2_bats.shape
```

```python
tels = []
tel_list = []
for f in freqs:
    if 348<f<352:
        tel = 'GBT350'
    elif 326<f<328:
        tel = 'AO327'
    elif 342<f<346:
        tel = 'WSRT350'
    elif 1357.8<f<1358.5 or 1449.8<f<1450.5:
        tel = 'AO1350'
    elif 1485<f<1580:
        tel = 'GBT1500'
    elif 1439<f<1440.1:
        tel = 'AO1440'
    elif 1330<f<1390:
        tel = 'WSRT1400'
    elif 810<f<830:
        tel = 'GBT820'
    else:
        tel = 'mystery'
    if tel not in tel_list:
        tel_list.append(tel)
    tels.append(tel_list.index(tel))
tels = np.array(tels)
for (i,t) in enumerate(tel_list):
    print t, np.sum(tels==i)
```

```python
for (i,t) in enumerate(tel_list):
    c = tels==i
    plt.plot(t_bats[c],
        (t2_bats[c]-t_bats[c])*86400*1e6, 
        ".", label=t)
plt.ylabel("T2 TOA minus T1 TOA ($\mu$s)")
plt.xlabel("MJD")
plt.legend(loc='best')
plt.savefig('tempo-baryerror.pdf')
```

```python
for (i,t) in enumerate(tel_list):
    c = tels==i
    plt.plot(t_bats[c],
        (t2_bats[c]-t2_bats[c].astype(float))*86400*1e6, 
        ".", label=t)
plt.ylabel("Difference ($\mu$s)")
plt.xlabel("MJD")
plt.legend(loc='best')
```
