```python
import warnings

import scipy.integrate
```

```python
G = 1. #AU, year, solar mass
```

```python
def acceleration(r):
    return -G*r/np.dot(r,r)**(3./2)
def deriv(rv, t):
    d = np.zeros_like(rv)
    d[:3] = rv[3:]
    d[3:] = acceleration(rv[:3])
    return d
```

```python
acceleration(np.array([0,1,0]))
```

```python
circular_initial_rv = np.array([1., 0, 0, 0, 1., 0])
eccentric_initial_rv = np.array([1,0,0,0,0.5,0])
```

```python
ts = np.linspace(0,1000,10000)
posns = scipy.integrate.odeint(deriv, circular_initial_rv, ts, atol=1e-10)
```

```python
def plot_orbits(ts, xs):
    plt.plot(xs[:,0], xs[:,1])
    plt.gca().set_aspect('equal')
```

```python
plot_orbits(ts,posns[:,:3])
```

```python
def plot_conserved(ts, xs, vs):
    plt.figure()
    rs = np.sqrt(np.sum(xs**2, axis=1))
    ss = np.sqrt(np.sum(vs**2, axis=1))

    plt.plot(ts,-rs**(-1) + 0.5*ss**2)

    plt.figure()
    plt.plot(ts,xs[:,0]*vs[:,1]-xs[:,1]*vs[:,0])
```

```python
class VelocityVerlet(object):
    def __init__(self, F, x0, v0, t0, stepsize):
        self.F = lambda x: F(x)
        self.x = x0
        self.t = t0
        self.v = v0
        self.stepsize = stepsize
        self.last_F = self.F(self.x)
        
    def step(self):
        v_half = self.v + 0.5*self.stepsize*self.last_F
        self.x = self.x + self.stepsize*v_half
        self.last_F = self.F(self.x)
        self.v = v_half + 0.5*self.stepsize*self.last_F
        self.t += self.stepsize
    
```

```python
t_stop = 1000
step_size = 1e-4
tally_size = 0.1

ts = [0]
xs = [eccentric_initial_rv[:3]]
vs = [eccentric_initial_rv[3:]]

s = VelocityVerlet(acceleration, xs[-1], vs[-1], ts[-1], step_size)

while ts[-1]<t_stop:
    s.step()
    if s.t - ts[-1]>tally_size:
        ts.append(s.t)
        xs.append(s.x)
        vs.append(s.v)

ts = np.array(ts)
xs = np.array(xs)
vs = np.array(vs)
```

```python
plot_orbits(ts, xs)
plot_conserved(ts, xs, vs)
```

```python
def jacobian(t, rv):
    J = np.zeros((6,6), dtype=rv.dtype)
    r = rv[:3]
    rl = np.dot(r,r)
    for i in range(3):
        J[i,3+i] = 1
        for j in range(3):
            J[3+i,j] = G*r[i]*r[j]*3*rl**(-5./2)
            if i==j:
                J[3+i,j] += -G*rl**(-3./2)
    return J
```

```python
class EulerWithDerivatives(object):
    def __init__(self, rhs, jac, t0, y0, derivs):
        self.rhs = rhs
        self.jac = jac
        self.t = t0
        self.y = np.array(y0)
        self.derivs = np.array(derivs)
        if derivs.shape[-1]!=len(self.y):
            raise ValueError("Derivative vectors must each be the same length as state vector")
    
    def step(self, stepsize):
        v = self.rhs(self.y, self.t)
        dv = self.jac(self.y, self.t)
        self.t += stepsize
        self.y += stepsize*v
        self.derivs += stepsize*np.dot(dv,derivs.T).T
        
```
