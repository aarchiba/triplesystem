```python
def make_par(px=1.4, pmra=3.56, pmdec=-3.9, solarn0=10., ephem="DE421"):
    return """
PSR              J0337+17    
RAJ      03:37:43.826099
DECJ      17:15:14.826651
PX                    {px}
PMRA                  {pmra}
PMDEC                {pmdec}
F0    365.9533436144258189  0  0.0000000000656904
F1      7.833539631670D-15  0  1.490959143049D-17
PEPOCH        56500.000000
DM               21.313000
SOLARN0              10.00
EPHEM             {ephem}
CLK               TT(TAI)
TZRMJD  56100.13622674904492
TZRFRQ            1379.999
TZRSITE                  j
""".format(**locals())
```

```python
tempo2_program = '/home/aarchiba/software/tempo2/tempo2/tempo2'
tempo2_dir = '/home/aarchiba/software/tempo2/t2dir'
timfile = '0337+17-scott-2013-08-29.tim'
def mjd_fromstring(s):
    i,f = s.split(".")
    i = int(i)
    f = np.float128("0."+f)
    return i+f
def get_bbats(par):
    parfile = "0337_temporary.par"
    with open(parfile,"wt") as f:
        f.write(par)
    outline = "OUTPUT {bat} \n"
    e = os.environ.copy()
    e['TEMPO2'] = tempo2_dir
    o = subprocess.check_output([tempo2_program,
            "-output", "general2", "-s", outline,
            "-f", parfile, timfile], env=e)
    bbats = []
    for l in o.split("\n"):
        if not l.startswith("OUTPUT"):
            continue
        bbats.append(mjd_fromstring(l.split()[1]))
    return np.array(bbats,dtype=np.float128)
```
