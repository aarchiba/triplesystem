```python
import kepler
import quad_integrate
```

```python
a = 3
pb = 2
eps1 = 0.3
eps2 = 0.4
i = 0.5
lan = 0.1
q = 1/2.
x_cm = np.zeros(3)
v_cm = np.zeros(3)
t0 = 0.
orbit = (a,pb,eps1,eps2,i,lan,q,x_cm,v_cm,t0)
```

```python
ts = np.linspace(0,10*pb,1000)
k_states = np.array([kepler.kepler_two_body(
    a,pb,eps1,eps2,i,lan,q,x_cm,v_cm,t0-t)[0] 
    for t in ts])
```

```python
initial_state = k_states[0]

states = quad_integrate.integrate_times(
    quad_integrate.KeplerRHS(),
    initial_state, ts,
    atol = 1e-16, rtol = 1e-16,
    initial_dt = pb/100.)
```

```python
states_fine = quad_integrate.integrate_times(
    quad_integrate.KeplerRHS(),
    initial_state, ts,
    atol = 1e-18, rtol = 1e-18,
    initial_dt = pb/100.)
```

```python
plt.plot(ts,states[:,2]-k_states[:,2])
plt.plot(ts,states_fine[:,2]-k_states[:,2])
plt.plot(ts,states[:,2]-states_fine[:,2])
plt.figure()
plt.plot(ts,states[:,2])
plt.plot(ts,k_states[:,2])
plt.figure()
ix = 2
plt.plot(ts,[kepler.inverse_kepler_two_body(s)[ix] 
    for s in states])
plt.plot(ts,[kepler.inverse_kepler_two_body(s)[ix] 
    for s in states_fine])
plt.figure()
plt.plot(ts[1:],np.diff([
    kepler.inverse_kepler_two_body(s)[-1] 
    for s in states])%pb)
#plt.plot(ts,[kepler.inverse_kepler_two_body(s)[ix] 
#    for s in states_fine])
```

I think this is actually the limits on the accuracy of the Keplerian parameterization, since it doesn't seem to improve much with tolerance. On the other hand, the way it increases with time is highly suspicious. It looks like the ODE solutions are converging towards something different from the Keplerian solution. In fact I think they're converging towards the solution for very slightly different initial conditions than those that correspond to the Keplerian solution (probably within roundoff).

In any case it seems clear we can hammer the error down to quite small, over this time span.

Ultimate limits - the harmonic oscillator
-----------------------------------------

Just how small can we make the error? It's tricky to test, since the truncation error and roundoff error build up over time and we're interested in fairly long-time behaviour, which means many steps. Machine epsilon for quad precision is 1e-34 or so, which is about as small as we can hope to make atol and rtol. If truncation error is proportional to atol or rtol, then we should be able to extrapolate down to an ultimate limit. We'll hit roundoff before then; how long before?

```python
t2 = np.linspace(0,1000*np.pi,1000).astype(np.float128)
s2 = quad_integrate.integrate_times(
        quad_integrate.HarmonicRHS(),
        [1,0], t2, atol=1e-20, rtol=1e-20)
```

```python
plt.plot(t2,s2[:,0]-np.cos(t2))
plt.plot(t2,s2[:,1]+np.sin(t2))
plt.plot(t2,np.cos(t2)**2+np.sin(t2)**2-1)
```

```python
initial_state = kepler.kepler_three_body(*orbit3)[0]
print "m_psr:",initial_state[6]
print "m_2:",initial_state[13]
print "m_3:",initial_state[20]
```

```python
ts = np.linspace(0,100*pb_i,1000).astype(np.float128)

def sim(hs):
    states = [quad_integrate.integrate_times(
         quad_integrate.KeplerRHS(),
         initial_state, ts,
         atol = h, rtol = h,
         initial_dt = pb_i/100.)
     for h in hs]
    i = np.argmin(hs)
    for j in range(len(hs)):
        if j==i:
            continue
        plt.plot(ts, states[j][:,2]-states[i][:,2], 
            label="%g" % hs[j])
    plt.legend(loc="best")
    
sim([1e-10,1e-11,1e-12])
```

```python
h = 1e-12
ts_long = np.linspace(0,1000*pb_i,100000).astype(
                                            np.float128)
s_long = quad_integrate.integrate_times(
         quad_integrate.KeplerRHS(),
         initial_state, ts_long,
         atol = h, rtol = h,
         initial_dt = pb_i/100.)
```

```python
k_long = []
for t, s in zip(ts_long,s_long):
    k = kepler.inverse_kepler_three_body(s) 
    k = list(k)
    for n in ['_i', '_o']:
        t0ix = kepler.three_body_parameters.index('t0'+n)
        pbix = kepler.three_body_parameters.index('pb'+n)

        t0 = k[t0ix]
        t0 += t
        t0 = (t0+0.5*k[pbix])%k[pbix]-0.5*k[pbix]
        k[t0ix] = t0
    k_long.append(k)
```

```python
ix = 8
n = 500
print len(ts_long), len(k_long)
plt.plot(ts_long[:n], [k[ix] for k in k_long[:n]])
plt.title("%s variation" % 
    kepler.three_body_parameters[ix])
plt.figure()
plt.plot(ts_long, [k[ix] for k in k_long])
plt.title("%s variation" % 
    kepler.three_body_parameters[ix])
```

```python
def show_param(name):
    ix = kepler.three_body_parameters.index(name)
    plt.figure()
    plt.plot(ts_long[:n], [k[ix] for k in k_long[:n]])
    plt.title("%s variation" % name)
    plt.savefig("%s-short.pdf" % name)
    plt.figure()
    plt.plot(ts_long, [k[ix] for k in k_long])
    plt.title("%s variation" % name)
    plt.savefig("%s-long.pdf" % name)

```

```python
show_param('pb_i')
show_param('pb_o')
```
