```python
delay_list = []
with open("0337_delays_2.txt") as f:
    for l in f.readlines():
        if l.startswith("#"):
            continue
        mjd, delay, tel = l.split()
        delay_list.append((float(mjd),float(delay),tel))
mjds = np.array([m for (m,d,t) in delay_list])
delays = np.array([d for (m,d,t) in delay_list])

tel_list = list(sorted(set([t for (m,d,t) in delay_list])))
tels = np.array([tel_list.index(t) for (m,d,t) in delay_list])

mjds -= delays/86400.

ix = np.argsort(mjds)
mjds = mjds[ix]
delays = delays[ix]

tels = tels[ix]
```

```python
def plot_residuals(choice=best_ix,
    xbins=20,
    ybins=24,
    shift = 0.):
    pos = minima[choice, 1:15]
    fs = threebody.compute_orbit(pos, mjds)
    r2 = threebody.remove_trend(fs['states'][:,2]-delays, mjds, tel_list, tels)
    print minima[choice, 0], np.sqrt(np.mean(r2**2))
    plt.cool()
    ns = kepler.three_body_parameters_measurable
    pb_i = pos[ns.index('pb_i')]
    pb_o = pos[ns.index('pb_o')]
    tasc_o = pos[ns.index('tasc_o')]
    tasc_i = pos[ns.index('tasc_i')]
    phases_i = ((mjds-tasc_i)/pb_i-shift)%1+shift
    wt, xedges, yedges = np.histogram2d(phases_i, mjds,
        [xbins,ybins], [[0+shift,1+shift],[np.amin(mjds),np.amax(mjds)]])
    res2d, xedges, yedges = np.histogram2d(phases_i, mjds,
        [xbins,ybins], [[0+shift,1+shift],[np.amin(mjds),np.amax(mjds)]],
        weights=np.asarray(r2*1e6,float))
    r = ma.array(res2d/wt)
    r[wt==0] = ma.masked
    plt.pcolormesh(xedges, yedges/365.2425, r.T) # FIXME: double-check orientation
    plt.ylim(np.amin(mjds)/365.2425,
        np.amax(mjds)/365.2425)
    plt.ylabel("time (y)")
    plt.xlabel("inner orbital phase")
    plt.xlim(shift, 1+shift)
    plt.colorbar(orientation="horizontal")

    plt.twinx()
    def mtoo(m):
        return (m-tasc_o)/pb_o+1
    plt.ylim(mtoo(np.amin(mjds)),
        mtoo(np.amax(mjds)))
    plt.ylabel("outer orbital phase")
```
