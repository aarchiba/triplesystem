```python
F = threebody.Fitter("0337+17-scott-2013-06-06",tzrmjd_middle='weighted',ppn_mode='GRtidal')
#F = threebody.Fitter("0337+17-scott-2013-06-06",tzrmjd_middle='weighted',
#    only_tels=['AO1440','AO1350','GBT1500'],
#    ppn_mode='heavysimple')
F_GR = threebody.Fitter("0337+17-scott-2013-06-06",tzrmjd_middle='weighted',ppn_mode='GR')
#F = threebody.Fitter("0337+17-scott-2013-06-06",tzrmjd_middle=True)
```

```python
print F.parameters
print len(F.parameters)
```

```python
M = minuit_wrap.Fitter(F.mfun)
```

```python
r_GR = F_GR.residuals(F_GR.best_parameters)
```

```python
r_both = F.residuals(F.best_parameters)
```

```python
p_tidal = F.best_parameters.copy()
p_tidal['Omega'] = 0
r_tidal = F.residuals(p_tidal)
```

```python
plt.hexbin(
        ((F.mjds-F.base_mjd-F.best_parameters['tasc_i'])/F.best_parameters['pb_i'])%1, 
        ((F.mjds-F.base_mjd-F.best_parameters['tasc_o'])/F.best_parameters['pb_o']),
        C=(r_tidal-r_GR)/F.best_parameters['f0']*1e6, gridsize=20)
plt.title(r"Tides only")
plt.ylabel("Outer phase")
plt.xlabel("Inner phase")
plt.colorbar().set_label(r"$\mu$s")
plt.savefig("tides-simple.pdf")

plt.figure()
plt.hexbin(
        ((F.mjds-F.base_mjd-F.best_parameters['tasc_i'])/F.best_parameters['pb_i'])%1, 
        ((F.mjds-F.base_mjd-F.best_parameters['tasc_o'])/F.best_parameters['pb_o']),
        C=(r_both-r_tidal)/F.best_parameters['f0']*1e6, gridsize=20)
plt.title(r"Rotation only")
plt.ylabel("Outer phase")
plt.xlabel("Inner phase")
plt.colorbar().set_label(r"$\mu$s")
plt.savefig("rotation-simple.pdf")
```
