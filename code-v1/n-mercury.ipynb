```python
a = 193.164
e = 0.205
pb = 87.969
xvm2 = kepler.kepler_two_body(a, pb, e, 0, np.pi/2, 0, (1.65956e-7)**(-1), np.zeros(3), np.zeros(3), 0)[0]
```

```python
xvm2.reshape((-1,7))
```

```python
xv = np.zeros(21)
xv[:14] = xvm2
xv[14] = 1e20
xv[17] = 1
xv[20] = 1e-40
```

```python
def periastron_advance(Pb, e, Mtot):
    return np.radians(39.73)/365.2425*(Pb*24)**(-5./3)*(1-e**2)**(-1)*Mtot**(2./3)
```

```python
newtrhs = quad_integrate.KeplerRHS(special=False, general=False)
```

```python
G = kepler.G
c = 86400
#cst = G**2*c**(-4)
#cst2 = G**3*c**(-6)
m0, m1, m2 = xv[6::7]
# can zap Theta terms with gamma=0
# can zap three-body terms with beta=1/2
#ppnrhs = quad_integrate.KeplerRHS(special=False, general=False, ppn_motion=True,
#    gamma=0, beta=0.5, c_scale=1.)
ppnrhs = quad_integrate.KeplerRHS(special=False, general=False, ppn_motion=True)
#cst,cst2
```

```python
newtrhs.evaluate(xv.astype(np.float128), 0).reshape((-1,7))
```

```python
ppnrhs.evaluate(xv.astype(np.float128), 0).reshape((-1,7))
```

```python
ppnrhs.evaluate(xv.astype(np.float128), 0).reshape((-1,7))[0,3]/newtrhs.evaluate(xv.astype(np.float128), 0).reshape((-1,7))[0,3]
```

```python
c
```

```python
states = []
t_bbs = []
t_psrs = []
t_ds = []
O = quad_integrate.ODEDelay(ppnrhs,
                            xv, 0,
                            rtol=1e-16, atol=1e-16)
for t in linspace(1,2000,20000):
    O.integrate_to(t)
    states.append(O.x)
    t_bbs.append(O.t_bb)
    t_psrs.append(O.t_psr)
    t_ds.append(O.t_d)
    
                            
```

```python
kepler_states = []
for s in states:
    kepler_states.append(kepler.inverse_kepler_two_body(s[:14]))
```

```python
oms = [(np.arctan2(k[2],k[3])) % (np.pi) for k in kepler_states]
plt.plot(t_ds,oms,label="simulated")
rate = 42.98*2*np.pi/(360*60*60)/(365.2425*100)
plt.plot(t_ds,[oms[0]+rate*t_d for t_d in t_ds],label="observed")
rate = 6*np.pi/pb*(2*np.pi*G*(m1+m0)/c**3/pb)**(2./3)*(1-e**2)**(-1)
plt.plot(t_ds,[oms[0]+rate*t_d for t_d in t_ds],label="formula")
#print pb, e, m1+m0
rate = periastron_advance(pb, e, m1+m0)
plt.plot(t_ds,[oms[0]+rate*t_d for t_d in t_ds],label="formula")
plt.legend(loc="best")
plt.xlabel("time (days)")
plt.ylabel("$\Omega$")
plt.savefig("periastron-advance.pdf")
```

```python
examples = [
    (193.164, 0.205, 87.969, (1.65956e-7)**(-1), "Mercury"),
    (4., 0.001, 0.198, 7.2, "J1023 (rough)"),
    (1.92, 1e-4, 1.6, 0.137, "J0337 (inner)"),
    (100, 0.1, 100, 1, "Random 1"),
    (100, 0.1, 10, 1, "Random 2"),
    ]

for i, (ai,ei,pbi,qi,namei) in enumerate(examples):
    plt.figure()
    xvm2i = kepler.kepler_two_body(ai, pbi, ei, 0, np.pi/2, 0, qi, np.zeros(3), np.zeros(3), pbi/4)[0]
    xvi = np.zeros(21)
    xvi[:14] = xvm2i
    xvi[16] = 1e20
    xvi[17] = 1
    xvi[20] = 1e-30

    statesi = []
    t_bbsi = []
    t_psrsi = []
    t_dsi = []
    Oi = quad_integrate.ODEDelay(ppnrhs,
                                 xvi, 0,
                                 rtol=1e-16, atol=1e-16)
    for t in linspace(0,50*pbi,10000):
        Oi.integrate_to(t)
        statesi.append(Oi.x)
        t_bbsi.append(Oi.t_bb)
        t_psrsi.append(Oi.t_psr)
        t_dsi.append(Oi.t_d)

    statesi = np.array(statesi)
    kepler_statesi = []
    for s in statesi:
        kepler_statesi.append(kepler.inverse_kepler_two_body(s[:14]))

    plt.subplot(141)
    omsi = [(np.arctan2(k[2],k[3])) % (np.pi) for k in kepler_statesi]
    plt.plot(t_dsi,omsi,label="simulated")

    m0i, m1i = xvm2i[6::7]
    print namei, m0i, m1i
    ratei = periastron_advance(pbi, ei, m1i+m0i)
    plt.plot(t_dsi,[omsi[0]+ratei*t_d for t_d in t_dsi],label="formula")
    plt.title(namei)
    plt.subplot(142)
    plt.plot(statesi[:,0],statesi[:,2])
    plt.gca().set_aspect('equal')
    plt.subplot(143)
    plt.plot(t_dsi,statesi[:,0])
    plt.plot(t_dsi,statesi[:,2])

    plt.subplot(144)
    plt.plot(t_dsi,statesi[:,3]/c)
    #plt.plot(t_dsi,statesi[:,4]/c)
    plt.plot(t_dsi,statesi[:,5]/c)

    plt.plot(t_dsi,statesi[:,10]/c)
    #plt.plot(t_dsi,statesi[:,11]/c)
    plt.plot(t_dsi,statesi[:,12]/c)

    plt.gcf().set_size_inches(8,2)
    plt.savefig(namei+"-periastron.pdf")

#print len(t_dsi), len(omsi), ratei, omsi[0]
#print np.amin(t_dsi), np.amax(t_dsi)
```
