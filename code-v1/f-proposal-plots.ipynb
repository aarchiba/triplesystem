```python
p_order = [
 'asini_i',
 'tasc_i',
 'pb_i',
 'eps1_i',
 'eps2_i',
 'asini_o',
 'tasc_o',
 'pb_o',
 'eps1_o',
 'eps2_o',
 'd_RAJ',
 'd_DECJ',
 'd_PMRA',
 'd_PMDEC',
 'd_PX',
 'acosi_i',
 'q_i',
 'acosi_o',
 'delta_lan',
 'lan',
 'dgamma',
 'dbeta',
 'delta',
 ]
if len(p_order)!=len(F.parameters):
    p_order = list(F.parameters)[:]
    i = p_order.index('delta')
    del p_order[i]
    p_order.append(F.parameters[i])
    print p_order
```

```python
J2 = J.copy()
for j, p in enumerate(p_order):
    J2[:,j] = J[:,F.parameters.index(p)]
Q, R = scipy.linalg.qr(J2, mode='economic')
```

```python
for p, d in zip(p_order, np.diag(R)):
    print p, d
```

```python
sim_r = Q[:,-1]*R[-1,-1]*delta_step/F.best_parameters['f0']
sim_r.min(), sim_r.max()
```

```python
b = J2[:,-1]*delta_step
A = J2[:,:-1]
x, res, rk, s = scipy.linalg.lstsq(A,b)
sim_r_2 = (b - np.dot(A,x))/F.best_parameters['f0']
sim_r_2.min(), sim_r_2.max()
```
