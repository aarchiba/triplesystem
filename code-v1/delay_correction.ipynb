Delay correction
================

Suppose at time $t$ we observe an excess delay of $z$. The thing is, this is actually $z(t-z)$, since it takes an extra time $z$ for the light to reach us. So let's define $\tau = t-z$; then $\tau$ is the natural time coordinate for the system evolution. We know $t$ increases monotonically with $\tau$; in fact since all our derivatives are  with respect to $\tau$, we have $dt/d\tau = 1 + v_z$. Can we turn this around? We have to be careful, because $d/dt = dt/d\tau d/d\tau = (1+v_z) d/d\tau$, so that for example $dz/dt = (1+v_z)v_z$.

Let's see if we can test this.

```python
import scipy.integrate
```

```python
def rhs_1(y, tau):
    [z, vz, t, tau_] = y
    return [vz, -z, 1+vz, 1]

taus = np.linspace(0,4*pi,1000)
r_1 = scipy.integrate.odeint(rhs_1, [0,0.5,0,0], taus)
```

```python
def rhs_2(y, t):
    [z, vz, t, tau_] = y
    a = np.array([vz, -z, 1+vz, 1])
    return (1+vz)**(-1)*a
ts = np.linspace(0,4*pi,1000)
r_2 = scipy.integrate.odeint(rhs_2, [0,0.5,0,0], ts)
```
