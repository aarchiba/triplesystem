PSR J0337+17 viewed face-on. Red indicates the outer companion, orange the inner, blue the pulsar, and gray the center of mass of the inner system. Objects are not to scale, but orbits are; the right panel is an enlargement of the inner orbit. Plotted here are the Keplerian orbits that are osculating at our reference epoch (MJD 55920), along with the body positions at that epoch. The actual orbit, including interactions, differs slightly but not enough to be visible on a diagram at this scale. Arrows indicate velocities; the velocity scale for each panel was chosen to make the arrows readable. Solid lines connect the center of mass to the periastron of each orbit, while dotted lines show the orientation of each eccentricity vector to allow comparison.  Our view of the system is from below with an inclination of approximately $39^\circ$. The tick mark on each orbit indicates the position of the ascending node, the moment where the pulsar passes through the plane of the sky moving away from us.
