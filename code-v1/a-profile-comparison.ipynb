```python
tel = 'WSRT'

if tel=='WSRT':
    stdprof_name = "data/wsrt/J0337+1715_1380MHz_MJD55900-56200_Jan2012-Sep2012.AR"
    fglob = "data/wsrt/*.MartenScr"
    ds_bins = 2
elif tel=='GBT':
    stdprof_name = "data/gbt/GUPPI_Lband_rotated.template"
    fglob = "data/gbt/*.forTOAs"
    ds_bins = 2
```

```python
#std_prof = psrchive.Archive_load("data/wsrt/J0337+17_1380_WSRT.std")
std_prof = psrchive.Archive_load(stdprof_name)
std_prof.pscrunch()
std_prof.fscrunch()
std_prof.tscrunch()
#std_prof.bscrunch_to_nbin(256)
std_np = std_prof.get_data()[0,0,0,:]
#std_np -= np.mean(std_np)
#std_np /= np.sqrt(np.sum(std_np**2))
plt.plot(np.linspace(-1,1,2*len(std_np),endpoint=False),
         np.concatenate([std_np,std_np]))
print len(std_np)
std_np_ds = profile_manipulation.downsample_by(std_np, ds_bins)
```

```python
# Tests of approx_fit
aa = 20
bb = -30
cc = 0.0001
phase, a, b, c, fit_prof = profile_manipulation.approx_fit(std_np, 
    aa*(std_np+cc*np.random.randn(len(std_np)))+bb,
    oversample=64)

print phase, 0
print a, aa
print b, bb
print c, cc
print np.mean(std_np), np.mean(fit_prof)
```

```python
files = []
for f in sorted(glob(fglob)):
    F = psrchive.Archive_load(f)
    if F.get_centre_frequency()<1000:
        continue
    F.pscrunch()
    F.fscrunch()
    #F.bscrunch_to_nbin(256)
    files.append(F)
print len(files)
```

```python
times, profiles, raw_profiles, noises = profile_manipulation.align_profiles(files,std_np,ds_by=ds_bins)
```

```python
plt.subplot(211)
plt.plot(std_np_ds)
plt.xlim(0,len(std_np_ds))
plt.subplot(212)
diffs = profiles-std_np_ds[None,:]
vmin, vmax = profile_manipulation.mostly_range(diffs)
plt.pcolormesh(diffs, vmin=vmin, vmax=vmax, rasterized=True)
plt.xlim(0,len(std_np_ds))
plt.subplots_adjust(hspace=0)
plt.cool()

plt.savefig("profile-diffs-ix-%s.pdf" % tel)
```

```python
start, end = 55920, 56500
b_profiles = profile_manipulation.bin_profiles(times,
    profiles-std_np_ds[None,:],
    n_bins=256,
    bin_range=(start,end),weights=1/noises)

xs = np.linspace(0,1,b_profiles.shape[1]+1)
plt.subplot(211)
plt.plot(xs[:-1],std_np_ds)
plt.subplot(212)
vmin, vmax = profile_manipulation.mostly_range(b_profiles)
plt.pcolormesh(xs,
               np.linspace(start,end,b_profiles.shape[0]+1),
               b_profiles, vmin=vmin, vmax=vmax,
               rasterized=True, shading='flat')
plt.ylim(start,end)
plt.subplots_adjust(hspace=0)
plt.cool()
plt.savefig("profile-diffs-mjd-%s.pdf" % tel)
```
