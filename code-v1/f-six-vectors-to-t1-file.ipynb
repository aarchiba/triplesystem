```python
outtimfile = "0337_6vec_table_GBT_Lband_fold_150131.tim"
V = astropy.table.Table.read("0337_6vec_table_GBT_Lband_fold_150131.fits")
```

```python
V
```

```python
imjds = V['tdbs'][:,0]-2400000.5
fmjds = V['tdbs'][:,1]
obscode = '1'
freqs = V['freqs']
errors = V['errors']
```

```python
with open(outtimfile, "wt") as f:
    for i in range(len(V)):
        imjd = str(int(imjds[i]))
        fmjd = ("%.20f" % fmjds[i])[2:]
        if len(fmjd)<13:
            fmjd += "0"*(13-len(fmjd))
        s = ("%s " % obscode
             +" "*14
             +"%8.3f" % freqs[i]
             +"%6s.%13s" % (imjd,fmjd[:13])
             +"% 9.3f" % errors[i])
        
        
        #print [s[15:24],s[24:44],s[44:53],s]
        f.write(s+"\n")
```
