```python
def f_one(x):
    return np.sin(x)+1e-6*np.random.randn()
def df_one(x):
    return np.cos(x)
```

```python
def deriv(f, x, h, vals=None):
    if vals is None:
        vals = 5*[None]
    if len(vals)%2==0:
        raise ValueError("Unable to work with even-length numbers of derivatives")
    for i in range(len(vals)):
        j = i-len(vals)//2
        if j==0:
            continue
        if vals[i] is None:
            vals[i] = f(x+j*h)
    if len(vals)==5:
        return (vals[0]-8*vals[1]+8*vals[3]-vals[4])/(12*h)
    elif len(vals)==3:
        return (-vals[0]+vals[2])/(2*h)
```
