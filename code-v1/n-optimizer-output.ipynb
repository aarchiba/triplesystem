```python
import os
from glob import glob

import threebody
```

```python
def load(jobid):
    job_dir = glob(os.path.join("optimizer-runs",jobid+".*"))[0]
    ps, chi2s = [], []
    for p,c2 in zip(sorted(glob(os.path.join(job_dir,"*-ps.npy"))),
                   sorted(glob(os.path.join(job_dir,"*-chi2s.npy")))):
        ps.append(np.load(p))
        chi2s.append(np.load(c2))
    return job_dir, np.array(ps), np.array(chi2s)
```

```python
jobid = "843"
job_dir, ps, chi2s = load(jobid)
print ps.shape, chi2s.shape
```

```python
F = threebody.Fitter("0337+17-scott-2013-06-06",tzrmjd_middle=True)
```

```python
l1 = 'acosi_i'
l2 = 'q_i'
i1 = F.parameters.index(l1)
i2 = F.parameters.index(l2)
plt.hexbin(ps[:,:,i1].reshape(-1),
           ps[:,:,i2].reshape(-1),
           C=chi2s.reshape(-1),
           reduce_C_function=np.amin)
best_p = ps.reshape((-1,ps.shape[-1]))[np.argmin(chi2s)]
plt.plot(best_p[i1],best_p[i2],'ko')
plt.plot(ps[-1,:,i1],ps[-1,:,i2],"k+")
```
