```python
#fpickle = "jac-telescope-params.pickle"
fpickle = "fitter_params.pickle"
d = pickle.load(open(fpickle,"rb"))
F = threebody.Fitter(**d)
F.parameters
```

```python
residuals = F.residuals(F.best_parameters)
```

```python
for i,t in enumerate(F.tel_list):
    c = F.tels==i
    r = residuals[c]
    u = F.phase_uncerts[c]
    
    plt.figure()
    plt.hist(r/u,bins=np.sqrt(len(r)),histtype="step",normed=True)
    xs = np.linspace(-10,10,1000)
    ys = scipy.stats.norm.pdf(xs)
    plt.plot(xs,ys)
    
    plt.plot(xs*np.std(r/u),ys/np.std(r/u))

    plt.title(t)
    plt.xlabel("z")
    plt.ylim(ymin=1e-4)
    plt.gca().set_yscale("log")
```

```python
raw_equads = {}
for i,t in enumerate(F.tel_list):
    c = F.tels==i
    r = residuals[c]
    u = F.phase_uncerts[c]
    
    qs = np.linspace(0,2*np.std(r),1000)
    ads = []
    for q in qs:
        zs = r/np.hypot(u,q)
        zs -= np.mean(zs)
        zs /= np.std(zs)
        A2, crit, sig = scipy.stats.anderson(zs.astype(float))
        ads.append(A2)
    plt.figure()
    plt.plot(qs,ads)
    raw_equads[t] = qs[np.argmin(ads)]
    plt.axvline(raw_equads[t])
    plt.xlabel("EQUAD (phase)")
    plt.ylabel("$A^2$")
    plt.title(t)
    plt.savefig("equad-AD-%s.pdf" % t)
    #print t, crit, sig
```

```python
equads = {}
efacs = {}
for i,t in enumerate(F.tel_list):
    c = F.tels==i
    r = residuals[c]
    u = F.phase_uncerts[c]
    efacs[t] = np.std(r/np.hypot(u,raw_equads[t]))
    equads[t] = raw_equads[t]*efacs[t]
    good_u = np.hypot(efacs[t]*u,equads[t])
    
    plt.figure()
    plt.hist(r/(u*np.std(r/u)),bins=np.sqrt(len(r)),
             histtype="step",normed=True,
             label="Scaled")
    plt.hist(r/good_u,bins=np.sqrt(len(r)),
             histtype="step",normed=True,
             label="EQUAD+EFAC")
    xs = np.linspace(-10,10,1000)
    ys = scipy.stats.norm.pdf(xs)
    plt.plot(xs,ys,
             label="Std. normal")
    plt.title(t)
    plt.xlabel("z")
    plt.ylabel("Probability density (per turn)")
    plt.ylim(ymin=1e-3)
    plt.gca().set_yscale("log")
    plt.text(0,1,"EQUAD %g turns\nEFAC %.3g" % (equads[t],efacs[t]),
             transform=plt.gca().transAxes,
             horizontalalignment="left",
             verticalalignment="top")
    plt.legend(loc="best")
    plt.savefig("equad-post-residuals-%s.pdf" % t)
```
