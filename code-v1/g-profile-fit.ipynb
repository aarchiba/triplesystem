```python
prof = A.get_data()[0,0,0,:]
xsl = np.linspace(0,1,1+len(prof))
xsm = (xsl[1:]+xsl[:-1])/2
plt.plot(xsm, prof, drawstyle="steps")
```

```python
def peak(x, w):
    kappa = (2*np.pi*w)**(-2)
    if kappa<7e2:
        rx = 2*np.pi*(xsl-x)
        d = scipy.stats.vonmises(kappa)
        r = d.cdf(rx[1:])-d.cdf(rx[:-1])
    else:
        rx = 2*np.pi*(xsl-x)
        d = scipy.stats.vonmises(kappa)
        r = d.cdf(rx[1:])-d.cdf(rx[:-1])
    r /= np.max(r)
    return r

plt.plot(xsm,peak(0.5,0.1))
plt.plot(xsm,peak(0.5,0.01))
plt.plot(xsm,peak(0.25,0.1))
plt.plot(xsm,peak(0.25,0.005))
```

```python
xx = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(xx, scipy.stats.vonmises(7e2).cdf(xx))
plt.plot(xx, scipy.stats.vonmises(7.5e2).cdf(xx))
plt.plot(xx, scipy.stats.vonmises_cython.von_mises_cdf_normalapprox(7.5e2,xx,50.1))
```

```python
def value_gaussians(gaussians):
    r = np.zeros_like(xsm)
    for (c, h, w) in gaussians:
        r += h*peak(c,w)
    return r
```

```python
gaussians = [(0.18, 11.5, 0.02),
             (0.325, 3, 0.025),
             (0.8, 9, 0.01),
             (0.74, 5.5, 0.03),
#             (0.75, 6, 0.01),
             (0.65, 2, 0.05),
             ]
plt.plot(xsm, prof)
plt.plot(xsm, value_gaussians(gaussians))
```
