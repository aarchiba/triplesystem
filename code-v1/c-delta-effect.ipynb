```python
mfun = F.make_mfun()
M = minuit_wrap.Fitter(F.make_mfun())
for p in F.parameters:
    M.values[p] = F.best_parameters[p]
    M.errors[p] = F.best_errors[p]
M.values['delta'] = new_p['delta']
M.set_normalization()
M.fixed['delta'] = True
```

```python
M.best_filename = "delta_best_values.pickle"
M.tol = 100
M.eps = 1e-6
M.strategy = 1
M.printMode = 3
M.migrad()
```

```python
M.best_values
```

```python
fit_p = new_p.copy()
for p in F.parameters:
    fit_p[p] = M.best_values[p]
rfit = F.residuals(fit_p)
```
