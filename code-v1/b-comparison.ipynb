```python
import os
from glob import glob

import kepler
import threebody
```

```python
datasets = [("All", 7983),
            ("AO", 11893),
            ("GBT", 11894),
            ("WSRT", 11895),
            ]
```

```python
parameters = None
pm0 = None
pe0 = None

r = []
plots = []
errors = []
for n, d in datasets:
    job_dir = glob(os.path.join("emcee-chains","%d.*"%d))[0]
    fp = np.load(os.path.join(job_dir,"fitted_parameters.npz"))
    pm = fp['parameter_means'][None][0]
    pe = fp['parameter_errors'][None][0]
    if parameters is None:
        parameters = sorted([k for k in pm.keys() if not k.startswith("j_")])
        r.append([""])
        for p in parameters:
            r.append([p])
        pm0 = pm
        pe0 = pe
        
        r[0].append(n)
        for i, p in enumerate(parameters):
            r[i+1].append(threebody.fmt(pm0[p], pe0[p]))
    else:
        r[0].append(n)
        v, e = [], []
        for i, p in enumerate(parameters):
            if p in pm:
                r[i+1].append(threebody.fmt((pm[p]-pm0[p])/pe0[p], pe[p]/pe0[p]))
                v.append((pm[p]-pm0[p])/pe0[p])
                e.append(pe[p]/pe0[p])
            else:
                v.append(0)
                e.append(0)
                r[i+1].append('*')
        plt.errorbar(range(len(v)),v,e, linestyle="none", label=n)
plt.legend(loc="best")
def func(i):
    try:
        return parameters[int(i)-1]
    except IndexError:
        return ""
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda i, j: func(i)))
plt.gca().xaxis.set_major_locator(IndexLocator(1,0))
plt.setp(plt.gca().get_xticklabels(), rotation=90)
#plt.gca().set_yscale("symlog", linthreshy=1)
plt.gcf().set_size_inches(10,10)
plt.gcf().tight_layout(pad=0.1)
plt.savefig("comparison.pdf")
plt.ylim(-30,30)
plt.savefig("comparison-zoom.pdf")

ws = [max(*[len(e[j]) for e in r]) for j in range(len(r[0]))]
with open("comparison.txt","wt") as f:
    for l in r:
        s = ""
        for e, w in zip(l,ws):
            s += ("%"+str(w+2)+"s")%e
        print s
        f.write(s)
        f.write("\n")
```
