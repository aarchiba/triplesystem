```python
from glob import glob
import os

import scipy.linalg

import pymc

import threebody
```

```python
F = threebody.Fitter()
```

```python
# Any distribution I put here is a prior, so let's be broad
@pymc.stochastic()
def offset(value=np.zeros(len(F.parameters))):
    return 0

@pymc.deterministic(trace=False)
def resids(offset=offset):
    params = F.best_parameters.copy()
    for p,o in zip(F.parameters, offset):
        params[p] += o
    return F.residuals(params)/F.phase_uncerts
efac = 1.3
result = pymc.Normal('result',
                     mu=resids,
                     tau=efac**(-2.),
                     value=np.zeros(len(F.mjds)),
                     trace=False,
                     observed=True)
```

```python
def load_job(jobid):
    r = []
    p = os.path.join("chains",jobid+"-*.*")
    for d in sorted(glob(p)+glob(os.path.join(p,"*/"))):
        try:
            db = pymc.database.txt.load(d)
        except ValueError:
            print d, "failed"
            continue
        if not db.trace_names:
            continue
        r.append(pymc.MCMC([offset,resids,result],db=db))
    return r
```

```python
r = load_job("726")
```

```python
def combine_offsets(r):
    o = []
    for m in r:
        o.append(m.trace('offset')[5000:])
    return np.concatenate(o)
```

```python
elev = 0.01
xs = np.linspace(-1,1,25000)
ys = xs+elev*np.random.randn(len(xs))

#m = pymc.Normal('m',mu=0,tau=(100000.)**(-2),value=1)
@pymc.stochastic()
def m(value=np.array([1,0])):
    return 0
@pymc.deterministic(trace=False)
def lresid(xs=xs, ys=ys, m=m):
    return m[0]*xs+m[1]-ys
rslt = pymc.Normal('rslt',mu=lresid,tau=elev**(-2),value=np.zeros(len(xs)),observed=True)

M2 = pymc.MCMC([m,lresid,rslt])
M2.use_step_method(pymc.AdaptiveMetropolis,m,cov=np.diag([1e-4,6e-5])**2)
M2.sample(10000)

```

```python
ms = M2.trace('m')[:]
print np.mean(ms,axis=0), np.std(ms[5000:],axis=0)
```
