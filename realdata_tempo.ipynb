```python
year_length = 365.2425
def remove_non_orbital(vec):
    non_orbital_basis = np.vstack([
                  np.cos(2*np.pi*mjds/year_length),
                  np.sin(2*np.pi*mjds/year_length),
                  mjds*np.sin(2*np.pi*mjds/year_length),
                  mjds*np.sin(2*np.pi*mjds/year_length),
                  np.cos(4*np.pi*mjds/year_length),
                  np.sin(4*np.pi*mjds/year_length),
                  mjds,
                  mjds**2/2.,
                  np.ones_like(mjds),
        ]).T
    x, res, rk, s = scipy.linalg.lstsq(non_orbital_basis, vec)
    return vec-np.dot(non_orbital_basis,x)
def remove_trend(vec):
    non_orbital_basis = np.vstack([
                  mjds,
                  mjds**2/2.,
                  np.ones_like(mjds),
                  np.arange(1,len(tel_list))[:,None]==tels[None,:],
                  np.cos(2*np.pi*mjds/year_length),
                  np.sin(2*np.pi*mjds/year_length),
        ]).T
    x, res, rk, s = scipy.linalg.lstsq(non_orbital_basis, vec)
    return vec-np.dot(non_orbital_basis,x)
```

Tempo-style iteration

```python
tol=1e-16
```

```python
def pack(args):
    return np.array(list(args[:-2]) + list(args[-2]) + list(args[-1]))
def unpack(args):
    return list(args[:-6])+[args[-6:-3],args[-3:]]
def eval_zs(args):
    args = unpack(args)
    print "eval:", args,

    initial_values, jac = kepler.kepler_three_body_measurable(*(args+[0]))
    vectors = jac.T # all of them, why not?
    
    O = quad_integrate.ODE(quad_integrate.KeplerRHS(),
           initial_values, 0,
           rtol = tol, atol = tol,
           vectors = vectors,
           delta = 1e-10)
    zs = []
    jac_all = []
    for t in mjds:
        O.integrate_to(t)
        zs.append(O.x[2])
        jac_all.append(O.dx[2,:])
    zs = np.array(zs)
    jac_all = np.array(jac_all)
    
    evec = remove_trend(zs-delays)
    for i in range(jac_all.shape[1]):
        jac_all[:,i] = remove_trend(jac_all[:,i])
    print np.sqrt(np.mean(evec**2))
    return evec, jac_all
```

Run this to initialize the iteration

```python
op = pack(orbital_parameters_fit_g2)
zs, jac = eval_zs(op)
vals = [(op, zs, jac)]
```

Run this to advance by a step

```python
op, zs, jac = vals[-1]
jac_detrend = np.copy(jac)[:,:14]
for i in range(jac_detrend.shape[1]):
    jac_detrend[:,i] = remove_trend(jac[:,i])
U,s,Vh = scipy.linalg.svd(jac_detrend, full_matrices=False)
print s
si = s**(-1)
si[s<1e-8*s[0]] = 0
d_op = -np.dot(Vh.T,np.dot(np.diag(si),np.dot(U.T,zs)))
print "op:", op[:14]
print "d_op:", d_op
print
n_op = np.copy(op)
n_op[:14] += d_op
zs, jac = eval_zs(n_op)
vals.append((n_op,zs,jac))
```
