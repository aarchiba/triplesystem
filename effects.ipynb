```python
def remove_trend(vec):
    return threebody.remove_trend(vec, mjds, tel_list, tels, uncerts)
```

```python
pnames = kepler.three_body_parameters_measurable[:14]
```

```python
def resids2d(vec, orbit, subset=None, gridsize=25, use_uncerts=False):
    plt.figure()
    plt.cool()
    times = orbit['times']
    tasc_i = orbit['parameters'][pnames.index('tasc_i')]
    pb_i = orbit['parameters'][pnames.index('pb_i')]
    tasc_o = orbit['parameters'][pnames.index('tasc_o')]
    pb_o = orbit['parameters'][pnames.index('pb_o')]
    phases = ((times-tasc_i)/pb_i) % 1
    if subset is None:
        subset = slice(None,None,None)
    if use_uncerts:
        plt.hexbin(phases[subset], times[subset], C=vec[subset]/uncerts[subset], gridsize=gridsize)
    else:
        plt.hexbin(phases[subset], times[subset], C=1e6*vec[subset], gridsize=gridsize)
    plt.xlim(0,1)
    plt.xlabel("Inner orbital phase")
    plt.ylim(np.amin(times), np.amax(times))
    plt.ylabel("MJD-55920")
    cb = plt.colorbar(orientation="horizontal")
    if use_uncerts:
        cb.set_label("Residual ($\sigma$)")
    else:
        cb.set_label("Residual ($\mu$s)")
    plt.twinx()
    plt.ylabel("Outer orbital phase")
    plt.ylim((np.amin(times)-tasc_o)/pb_o+1, 
             (np.amax(times)-tasc_o)/pb_o+1)
```

```python
def resids1d(vec, orbit, subset=None):
    plt.figure()
    times = orbit['times']
    tasc_i = orbit['parameters'][pnames.index('tasc_i')]
    pb_i = orbit['parameters'][pnames.index('pb_i')]
    tasc_o = orbit['parameters'][pnames.index('tasc_o')]
    pb_o = orbit['parameters'][pnames.index('pb_o')]
    phases = ((times-tasc_i)/pb_i) % 1
    if subset is None:
        subset = np.ones(len(times),bool)
    for i,n in list(enumerate(tel_list))[::-1]:
        s = subset & (tels==i)
        plt.subplot(211)
        plt.plot(times[s], 1e6*vec[s], ".", label=tel_list[i])
        plt.subplot(212)
        plt.plot(phases[s], 1e6*vec[s], ".", label=tel_list[i])
        
    plt.subplot(211)
    plt.xlabel("MJD-55920")
    plt.ylabel("Residual ($\mu s$)")
    #plt.legend(loc="best")
    nx,xx = plt.gca().get_xlim()
    plt.twiny()
    plt.xlim((nx-tasc_o)/pb_o + 1,
             (xx-tasc_o)/pb_o + 1)
    plt.xlabel("Outer orbital phase")
    
    plt.subplot(212)
    plt.xlim(0,1)
    plt.xlabel("Inner orbital phase")
    plt.ylabel("Residual ($\mu s$)")
```

```python
def scale(p):
    return (np.array(p,np.float128)-np.array(orbital_parameters_fit,np.float128))/np.array(errors,np.float128)
def unscale(p):
    return np.array(p,np.float128)*np.array(errors,np.float128)+np.array(orbital_parameters_fit,np.float128)
```

```python
cache = {}
```

```python
def mfun(asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan):
    p = (asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan)
    p = tuple(unscale(p))
    if p in cache:
        v = cache[p]
    else:
        o = threebody.compute_orbit(p, mjds, shapiro=True, special=True, general=True)
        r = threebody.remove_trend(
            o['states'][:,2]+o['shapiro']+86400*o['einstein']-delays, mjds, tel_list, tels)
        v = np.sum((r/uncerts)**2)
        cache[p] = v
    return v
```

```python
m = minuit.Minuit(mfun)
```

```python
for n,v in zip(m.parameters, scale(orbital_parameters_fit)):
    m.values[n] = v
```

```python
m.printMode = 3
m.strategy = 2
m.eps = 1e-10
m.tol = 100.
m.up = 1.
m.migrad()
print m.fval/len(mjds)
```

```python
m.hesse()
```

```python
fit_fixed_shapiro = unscale([m.values[n] for n in m.parameters])
errors_fixed_shapiro = [m.errors[n]*e for n,e in zip(m.parameters,errors)]
for n,v,e in zip(m.parameters,fit_fixed_shapiro,errors_fixed_shapiro):
    print n, threebody.fmt(v,e)
print
C = m.matrix(True)
for i in range(len(m.parameters)):
    for j in range(i+1,len(m.parameters)):
        print m.parameters[i], m.parameters[j], C[i][j]
    print
```

```python
orbit_fixed = threebody.compute_orbit(fit_fixed_shapiro, mjds,
            shapiro=True, special=True, general=True)
```

```python
def plots(orbit, name):
    r = orbit['states'][:,2].copy()
    if 'shapiro' in orbit:
        r += orbit['shapiro']
    if 'einstein' in orbit:
        r += 86400*orbit['einstein']
    r = remove_trend(r-delays)
    resids2d(r, orbit)
    plt.gcf().set_size_inches(two_column, two_column)
    plt.savefig("hexresids_%s.pdf" % name)
    resids1d(r, orbit)
    plt.gcf().set_size_inches(10.5, 8)
    plt.savefig("resids1d_%s.pdf" % name)
```
