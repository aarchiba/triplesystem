```python
def remove_trend(vec):
    return threebody.remove_trend(vec, mjds, tel_list, tels)
```

```python
pnames = kepler.three_body_parameters_measurable[:14]
```

```python
def resids2d(vec, orbit, subset=None, gridsize=25):
    plt.figure()
    plt.cool()
    times = orbit['times']
    tasc_i = orbit['parameters'][pnames.index('tasc_i')]
    pb_i = orbit['parameters'][pnames.index('pb_i')]
    tasc_o = orbit['parameters'][pnames.index('tasc_o')]
    pb_o = orbit['parameters'][pnames.index('pb_o')]
    phases = ((times-tasc_i)/pb_i) % 1
    if subset is None:
        subset = slice(None,None,None)
    plt.hexbin(phases[subset], times[subset], C=1e6*vec[subset], gridsize=gridsize)
    plt.xlim(0,1)
    plt.xlabel("Inner orbital phase")
    plt.ylim(np.amin(times), np.amax(times))
    plt.ylabel("MJD-55920")
    cb = plt.colorbar(orientation="horizontal")
    cb.set_label("Residual ($\mu$s)$")
    plt.twinx()
    plt.ylabel("Outer orbital phase")
    plt.ylim((np.amin(times)-tasc_o)/pb_o+1, 
             (np.amax(times)-tasc_o)/pb_o+1)
```

```python
resids2d(remove_trend(orbit['states'][:,2]+orbit['shapiro']+86400*orbit['einstein']-delays), orbit)
plt.gcf().set_size_inches(two_column, two_column)
plt.savefig("hexresids.pdf")
```

```python
resids = remove_trend(orbit['states'][:,2]+orbit['shapiro']+86400*orbit['einstein']-delays)
for i,t in enumerate(tel_list):
    subset = tels==i
    resids2d(resids, orbit, subset)
    plt.title("%s (%f $\mu$s)" % (t, 1e6*np.std(resids[subset])))
    plt.gcf().set_size_inches(two_column, two_column)
    plt.savefig("resids-%s.pdf" % t)
```

```python
resids2d(remove_trend(orbit['shapiro']),orbit)
plt.title("Shapiro delay")
plt.gcf().set_size_inches(two_column, two_column)
plt.savefig("shapiro.pdf")

resids2d(remove_trend(86400*orbit_no_general['einstein']),orbit)
plt.title("Delay due to $\gamma$ (special only, detrended)")
plt.gcf().set_size_inches(two_column, two_column)
plt.savefig("einstein-special.pdf")

resids2d(remove_trend(86400*(orbit['einstein'] - orbit_no_general['einstein'])),orbit)
plt.title("Delay due to $\gamma$ (general only, detrended)")
plt.gcf().set_size_inches(two_column, two_column)
plt.savefig("einstein-general.pdf")

resids2d(remove_trend(86400*orbit['einstein']),orbit)
plt.title("Delay due to $\gamma$ (special and general, detrended)")
plt.gcf().set_size_inches(two_column, two_column)
plt.savefig("einstein.pdf")
```

```python
def scale(p):
    return (np.array(p,np.float128)-np.array(orbital_parameters_fit,np.float128))/np.array(errors,np.float128)
def unscale(p):
    return np.array(p,np.float128)*np.array(errors,np.float128)+np.array(orbital_parameters_fit,np.float128)
```

```python
assumed_uncert = 2.45e-6
```

```python
cache = {}
```

```python
def mfun(asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan):
    p = (asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan)
    p = tuple(unscale(p))
    if p in cache:
        v = cache[p]
    else:
        o = threebody.compute_orbit(p, mjds, shapiro=True, special=True, general=True)
        r = threebody.remove_trend(
            o['states'][:,2]+o['shapiro']+86400*o['einstein']-delays, mjds, tel_list, tels)
        v = np.sum((r/assumed_uncert)**2)
        cache[p] = v
    return v
```

```python
m = minuit.Minuit(mfun)
```

```python
m.printMode = 3
m.strategy = 2
m.eps = 1e-10
m.tol = 100.
m.migrad()
print assumed_uncert*(m.fval/len(mjds))**2
```

```python
fit_fixed_shapiro = unscale([m.values[n] for n in m.parameters])
errors_fixed_shapiro = [m.errors[n]*e for n,e in zip(m.parameters,errors)]
for n,v,e in zip(m.parameters,fit_fixed_shapiro,errors_fixed_shapiro):
    print n, threebody.fmt(v,e)
print
C = m.matrix(True)
for i in range(len(m.parameters)):
    for j in range(i+1,len(m.parameters)):
        print m.parameters[i], m.parameters[j], C[i][j]
    print
```
