```python
new_base = dict(new_RAJ=43.82589000,
                new_DECJ=14.8281000,
                new_PX=1.34,
                new_PMRA=0,
                new_PMDEC=0)
steps = dict(new_RAJ=0.0001,
                new_DECJ=0.001,
                new_PX=1,
                new_PMRA=0.01,
                new_PMDEC=0.01)
par_base = """
PSR              J0337+17    
RAJ      03:37:{new_RAJ} 
DECJ      17:15:{new_DECJ} 
PX                    {new_PX} 
PMRA                  {new_PMRA} 
PMDEC                 {new_PMDEC} 
F0    365.9533436144258189  0  0.0000000000656904
F1      7.833539631670D-15  0  1.490959143049D-17
PEPOCH        56100.000000
START            55930.923
FINISH           56293.906
DM               21.313000
SOLARN0              10.00
EPHEM             DE421
CLK               TT(TAI)
TZRMJD  56100.13622674904492
TZRFRQ            1379.999
TZRSITE                  j
"""
                
```

```python
def mjd_fromstring(s):
    i,f = s.split(".")
    i = int(i)
    f = np.float128("0."+f)
    return i+f

def get_bbats(new):
    p = par_base.format(**new)
    print p
    parfile = "0337_temporary.par"
    with open(parfile,"wt") as f:
        f.write(p)
    outline = "OUTPUT {bat} \n"
    o = subprocess.check_output([tempo2_program,
            "-output", "general2", "-s", outline,
            "-f", parfile, timfile], env=e)
    bbats = []
    for l in o.split("\n"):
        if not l.startswith("OUTPUT"):
            continue
        bbats.append(mjd_fromstring(l.split()[1]))
    return np.array(bbats,dtype=np.float128)
```

```python
bat_base = get_bbats(new_base)
```

```python
pn = "new_PMRA"
plus = new_base.copy()
plus[pn] += steps[pn]
bat_plus = get_bbats(plus)
```

```python
scales = dict(d_RAJ=-0.00725, d_DECJ=-4.85e-3, d_PX=-1000000, d_PMRA=-1.4e-7, d_PMDEC=-1.4e-7)
```

```python
plt.plot(bat_base,(bat_plus-bat_base)*86400*1e6,".",markersize=1)
d_name = "d_"+pn.split("_")[1]
plt.plot(bat_base, np.array(derivs[d_name])*scales[d_name],".",markersize=1)
#plt.axvline(56100)
#plt.axhline(0)
plt.title(r"$\Delta$ %s = %g" % (pn,steps[pn]))
plt.ylabel(r"$\mu$s")
plt.xlabel("MJD")
```
