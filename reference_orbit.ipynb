```python
import scipy.optimize

import minuit

import kepler
import quad_integrate
import threebody
```

```python
(asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
            acosi_i, q_i,
            asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
            acosi_o, delta_lan) = threebody.best_parameters
```

```python
n_per_inner = 100
n_outer = 3
times = np.arange(n_outer*pb_o/pb_i*n_per_inner)*(pb_i/n_per_inner)
```

```python
o = threebody.compute_orbit(threebody.best_parameters, times)
```

```python
params = []
for i in range(len(times)):
    op = kepler.inverse_kepler_three_body_measurable(o["states"][i],o["times"][i])[:14]
    params.append(op)
    if not i%1000:
        print i, len(times)
```

```python
o["parameter_names"] = kepler.three_body_parameters_measurable
o["parameters"] = np.array(params)
```

```python
np.savez("reference_orbit.npz", **o)
```

```python
s = o["states"]
cm_i = (s[:,:6]*s[:,6,None]+s[:,7:13]*s[:,13,None])/(s[:,6,None]+s[:,13,None])

plt.plot(o["times"],cm_i[:,2],"b")
plt.ylabel("z (lt-s)")
plt.xlabel("MJD-start")
plt.twinx()
plt.plot(o["times"],cm_i[:,5],"g")
plt.ylabel("v_z (lt-s/day)")
```

```python
delta_i = []
for i in range(len(o["times"])):
    s = o["states"][i]
    j_inner = np.cross(s[:3]-s[7:10],s[3:6]-s[10:13])
    cm_i = (s[6]*s[:6]+s[13]*s[7:13])/(s[6]+s[13])
    j_outer = np.cross(s[14:17]-cm_i[:3],s[17:20]-cm_i[3:6])
    j_inner /= np.sqrt(np.dot(j_inner,j_inner))
    j_outer /= np.sqrt(np.dot(j_outer,j_outer))
    delta_i.append(np.arccos(np.dot(j_inner,j_outer)))
o["delta_i"] = delta_i
```

```python
i_inv = []
for i in range(len(o["times"])):
    s = o["states"][i][:21].reshape(3,7)
    J = np.zeros(3)
    for j in range(3):
        J += s[j,6]*np.cross(s[j,:3],s[j,3:6])
    i_inv.append(np.arccos(J[2]/np.sqrt(np.dot(J,J))))
o["i_inv"] = i_inv
```

```python
omega_i = []
omega_o = []
for i in range(len(o["times"])):
    p = o["parameters"][i]
    omega_i.append(np.arctan2(
        p[o["parameter_names"].index("eps2_i")],
        p[o["parameter_names"].index("eps1_i")]))
    omega_o.append(np.arctan2(
        p[o["parameter_names"].index("eps2_o")],
        p[o["parameter_names"].index("eps1_o")]))
o["omega_i"] = omega_i
o["omega_o"] = omega_o
```

```python
phase_i = []
phase_o = []
for i in range(len(o["times"])):
    tasc_i = o["parameters"][i][o["parameter_names"].index("tasc_i")]
    tasc_o = o["parameters"][i][o["parameter_names"].index("tasc_o")]
    pb_i = o["parameters"][i][o["parameter_names"].index("pb_i")]
    pb_o = o["parameters"][i][o["parameter_names"].index("pb_o")]
    
    phase_i.append((o["times"][i]-tasc_i)/pb_i % 1)
    phase_o.append((o["times"][i]-tasc_o)/pb_o % 1)

o["phase_i"] = phase_i
o["phase_o"] = phase_o
```
