```python
data_0825=pickle.load(open('systematics-package-lband-20170825.pickle', 'rb'))
```

```python
par_dict=data_0825['best_parameters']
p_period = 0.00273258863228
pb_i = par_dict['pb_i']
pb_o = par_dict['pb_o']
phi_i=(2.0*math.pi)/pb_i
phi_o=(2.0*math.pi)/pb_o
tasc_i = par_dict["tasc_i"]
tasc_o = par_dict["tasc_o"]

my_data=data_0825
base_mjd=my_data['base_mjd']

limiit=10000
c = abs(my_data['residuals']/my_data['phase_uncerts'])<limiit
new_phase = np.array(my_data['residuals'][c], dtype=np.float64)
new_times = np.array(my_data['times'][c], dtype=np.float64)+base_mjd
new_unc = np.array(my_data['phase_uncerts'][c], dtype=np.float64)

print my_data['telescope_list']
print base_mjd
```

#Do additional fit:

```python
import arrow_plot as ap

d=np.array(my_data['derivatives'])[np.newaxis][0]
cols= sorted(d.keys())
A = np.array([d[c] for c in cols]).T
b=new_phase
r = ap.lstsq_with_errors(A, b, new_unc)
better_phase=b-np.dot(A,r.x)
diff=new_phase-better_phase

name=['WSRT_1400','GBT_1400', 'AO_1400']
color=['y', 'g', 'b']

rd.plot_res(my_data, new_times, better_phase, new_unc, name, color)
```
