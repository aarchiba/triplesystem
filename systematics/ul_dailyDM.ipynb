```python

def der_fit_new(data,phase,mjd,unc, sequence, res_out=False):

    d=np.array(data['derivatives'])[np.newaxis][0]
    
    d['cos12'] = np.cos(-1.*phi_o*(mjd)+2.*phi_i*(mjd))
    d['sin12'] = np.sin(-1.*phi_o*(mjd)+2.*phi_i*(mjd))
    
    apj= ap.draw_plot(par_dict, 4, mjd, d['cos12'], unc)
    
    cols= sorted(sequence)
    
    A=np.array([d[c] for c in cols]).T
    
    b=phase
    r = ap.lstsq_with_errors(A, b, unc)
    better_phase=b-np.dot(A,r.x)
    
    for (i,c) in enumerate(cols):
        if c in data['best_parameters']:
            v = data['best_parameters'][c]
        elif c in data['linear_parameters']:
            v = 0
        else:
            v=0.0#raise ValueError('mistery column %s'%c)
        print i, c, v, v-r.x[i], "+/-", r.uncerts[i], (r.x/r.uncerts)[i], "sigma"
        
    if res_out is False:
        return better_phase, r.singular_values
    else:
        return better_phase, A, r.x, cols
```
