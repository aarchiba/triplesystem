#Arrow lenght:

###functions:

```python
p_period = 0.00273258863228

def arrow_length(coeff):
    my_coeff=np.concatenate((coeff.U, coeff.V), axis=0)
    res=0.
    for k in range(0,len(my_coeff)):
        my_coeff[k]=my_coeff[k]*p_period*1e9
        res=res+(my_coeff[k]*my_coeff[k])
    res=res/(len(my_coeff))
    return np.sqrt(res)

def arrow_length_array(array):
    list_res=[]
    for i in range(0, len(array)):
        res=arrow_length(array[i])
        list_res.append(res)
    return list_res
```

```python
def get_arrow(x_range,y_range, my_ampl, par_dict, mjd, unc, my_dof, der=False, my_data=None):
    #make a grid:
    X, Y, U, V = ap.generate_xy([x_range[0],x_range[1]],[y_range[0],y_range[1]])
    #combine them into the object:
    my_coeff = ap.Coefficients(X, Y, U*my_ampl, V*my_ampl)
    #make an mjd array out of generated coefficients:
    my_arrows=ap.fake_arrow_array(my_coeff,par_dict,mjd)
    in_coeff=ap.ar_coeff(par_dict, 4, mjd, my_arrows, unc)
    if der==True:
        if my_data==None:
            print 'Error! Need data for derivatives!'
        else:
            data=my_data
            new_arrows=ap.der_fit(data,my_arrows,unc)
    else:
        new_arrows=my_arrows
    #fit the arrows to generated array:
    out_coeff=ap.ar_coeff(par_dict, 4, mjd, new_arrows, unc)
    #get the std of the arrow lengths
    doff=(len(out_coeff.X)-len(my_coeff.X))
    dof=my_dof+doff
    if dof<0:
        dof=0
    my_std=ap.get_arrow_length(out_coeff,dof)
    #diff_coeff=in_coeff-out_coeff
    return my_std, out_coeff#, diff_coeff
```
