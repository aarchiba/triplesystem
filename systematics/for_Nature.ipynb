```python
def coeff_plot(coeff, scale=None, plot_unc=True, color=err_bl, units='', lentext=True, note=False):
    err_grey=np.array([149, 165, 166])/255.
    mpl.rcParams['font.size'] = 7
    mpl.rcParams['font.family'] = 'sans-serif'
    mpl.rcParams['axes.linewidth'] = 1
    mpl.rcParams['ps.useafm'] = True
    mpl.rcParams['pdf.use14corefonts'] = True
    mpl.rcParams['text.usetex'] = True
    mpl.rcParams['ytick.labelsize']=7
    mpl.rcParams['xtick.labelsize']=7
    mm_to_in = 0.0393701 
    
    fig, ax = plt.subplots()

    if scale is None:
        ar_scale=np.amax(np.hypot(coeff.V,coeff.U))
    else:
        ar_scale=scale
    if plot_unc:
        for k in range(0,len(coeff.err_X)):
            i,j = coeff.err_ix[k]
            ax.plot(coeff.err_X[k]/ar_scale+j,coeff.err_Y[k]/ar_scale+i, color=color, lw=1, alpha=0.3)
    Q = plt.quiver(coeff.X, coeff.Y, coeff.U, coeff.V, units='x', scale=ar_scale, color=color)
    Acols=np.amax(coeff.X)+1
    ax.set_xlim(-Acols+0.05,Acols-0.05)
    ax.set_ylim(-0.95,Acols-0.05)
    QQ = plt.quiver(-3, 0, 50, 0.0, units='x', scale=ar_scale, color=err_grey, alpha=0.8)
    ax.set_xlabel('Inner orbital harmonic \n (multiple of $f_{b,inner}$)')
    ax.set_ylabel(r'Outer orbital harmonic \n (multiple of $f_{b,outer}$)')
    ax.set_aspect('equal')
    err_red=np.array([191, 54, 12])/255.0
    if lentext:
        ax.figtext(0.15,0.2, r'the longest arrow = %f %s'%(ar_scale,units), color=err_red)
    if note==True:
        ax.text(-3,0.1, r'50 ns'%(ar_scale), color=err_grey, fontsize=7, )
    return ar_scale


```

```python
def fit_plot(emac, number, name, color, numplots, limm):
    
    mpl.rcParams['font.size'] = 7
    mpl.rcParams['font.family'] = 'sans-serif'
    mpl.rcParams['axes.linewidth'] = 1
    mpl.rcParams['ps.useafm'] = True
    mpl.rcParams['pdf.use14corefonts'] = True
    mpl.rcParams['text.usetex'] = True
    mpl.rcParams['ytick.labelsize']=7
    mpl.rcParams['xtick.labelsize']=7
    mpl.rcParams['axes.labelsize']=7
    mm_to_in = 0.0393701
    
    majorLocator = MultipleLocator(50)
    majorFormatter = FormatStrFormatter('%.0f')
    minorLocator = MultipleLocator(10)
    yearLocator = MultipleLocator(1)

    #ax1.set_xlabel('Outer orbital phase')
    #ax1.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    #ax1.xaxis.set_major_formatter(majorFormatter)
    
    
    err_clr, face_clr, mec_clr=rd.color_fun(color)
    mu, std = norm.fit(emac)
    ax=plt.subplot(2, int(numplots/2), number)
    ax.hist(emac, bins=100, normed=True, alpha=0.9, facecolor=err_clr, edgecolor=err_clr, histtype='stepfilled')#, rwidth=0.1)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    ax.plot(x, p, 'k', linewidth=1)
    #ax.set_xlabel(r'$\mu$ = %.4f,  $\sigma$ = %.4f'%(mu, std))
    ax.set_xlim(-limm,limm)
    ax.yaxis.set_ticklabels([])
    ax.yaxis.set_ticks([])
    ax.xaxis.set_ticks_position('bottom')
    if numplots == 4:
        if number%2:
            xx=0.137
        else:
            xx=0.557
        if (number >2.5):
            ax.set_xlabel('residual/uncertainty')
            yy=0.45
        else:
            yy=0.87
    if numplots == 6:
        if number%3:
            if number%2:
                 xx=0.137
            else:
                 xx=0.407
        else:
            xx=0.707

        if (number >3.5):
            yy=0.45
            if number==4:
                 xx=0.137
            if number==5:
                 xx=0.407
        else:
            yy=0.87

    plt.figtext(xx,yy, '%s'%name, color=mec_clr)
    plt.figtext(xx+0.2,yy, r'$\mu$ = %.3f'%mu)
    plt.figtext(xx+0.2,yy-0.02, r'$\sigma$ = %.3f'%std)
    return std
```
