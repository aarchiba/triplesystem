```python
err_red=np.array([191, 54, 12])/255.0
err_bl=np.array([21, 101, 192])/255.0
err_green=np.array([85, 139, 47])/255.0
err_violet=np.array([49, 27, 146])/255.0
```

```python
def coeff_plot(coeff, scale=None, plot_unc=True, color=err_bl, units='', lentext=True, note=False):
    err_grey='k'
    mpl.rcParams['font.size'] = 7
    mpl.rcParams['font.family'] = 'sans-serif'
    mpl.rcParams['axes.linewidth'] = 1
    mpl.rcParams['ps.useafm'] = True
    mpl.rcParams['pdf.use14corefonts'] = True
    #mpl.rcParams['text.usetex'] = True
    mpl.rcParams['ytick.labelsize']=7
    mpl.rcParams['xtick.labelsize']=7
    mpl.rcParams['axes.unicode_minus']=False
    mm_to_in = 0.0393701 
    
    fig, ax = plt.subplots()

    if scale is None:
        ar_scale=np.amax(np.hypot(coeff.V,coeff.U))
    else:
        ar_scale=scale
    if plot_unc:
        for k in range(0,len(coeff.err_X)):
            i,j = coeff.err_ix[k]
            ax.plot(coeff.err_X[k]/ar_scale+j,coeff.err_Y[k]/ar_scale+i, color=color, lw=1, alpha=0.3)
    Q = plt.quiver(coeff.X, coeff.Y, coeff.U, coeff.V, units='x', scale=ar_scale, color=color)
    Acols=np.amax(coeff.X)+1
    ax.set_xlim(-Acols+0.05,Acols-0.05)
    ax.set_ylim(-0.95,Acols-0.05)
    QQ = plt.quiver(-3, 0, 50, 0.0, units='x', scale=ar_scale, color=err_grey, alpha=0.8)
    ax.set_xlabel('Inner orbital harmonic \n (multiple of $f_{b,inner}$)')
    ax.set_ylabel(r'Outer orbital harmonic \n (multiple of $f_{b,outer}$)')
    ax.set_aspect('equal')
    err_red=np.array([191, 54, 12])/255.0
    if lentext:
        ax.figtext(0.15,0.2, r'the longest arrow = %f %s'%(ar_scale,units), color=err_red)
    if note==True:
        ax.text(-3,-0.3, r'50 ns'%(ar_scale), color=err_grey, fontsize=7, )
    return ar_scale
```

#Residuals plot:

```python
import matplotlib.font_manager as fm
import matplotlib.gridspec as gridspec
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

err_w1400,face_w1400,mec_w1400 = rd.color_fun('y')
err_gbt,face_gbt,mec_gbt = rd.color_fun('g')
err_a1400,face_a1400,mec_a1400 = rd.color_fun('b')

mjd_year = (55197-55920)+base_mjd
year_mjd = 2010
def day_to_year(mjd):
    return (mjd-mjd_year)/365.2425+year_mjd

def day_to_tasc(mjd, base_mjd, tasc_o, pb_o):
    zero_tasc=tasc_o+base_mjd# - pb_o
    tasc_time=(mjd-zero_tasc)/pb_o
    return tasc_time


scl_mu=1e6*p_period
```

```python
def fig_res(width, hight, my_ms=1.0,my_mew='0.5',
            x1=0.08, y1=0.09, l1=0.82, h1=0.26, l2=0.024, gap=0.025,
            x_bin=67, y_bin=18, fig_name='res_fig'):

    import matplotlib as mpl

    scl_mu=1e6*p_period
    scl_ns=1e9*p_period
    deltaa=2.6e-6#2 sigma upper limit
    delta_only=ap.der_of_par(my_data,'delta',new_unc)

    my_cmap='RdBu'
    hex_edge=np.array([215, 219, 221])/255.

    fig = plt.figure()

    rect1 = [x1, y1+2.*(h1+gap), l1, h1]
    rect2 = [x1, y1+h1+gap, l1, h1]
    rect3 = [x1, y1, l1, h1]

    rect4 = [x1+l1, y1+h1+gap, l2, h1]
    rect5 = [x1+l1, y1, l2, h1]

    ax1 = fig.add_axes(rect1)
    ax2 = fig.add_axes(rect1)
    ax2 = ax1.twiny()
    ax3 = fig.add_axes(rect2, sharex=ax1) 
    ax4 = fig.add_axes(rect3)
    ax5 = fig.add_axes(rect4)
    ax6 = fig.add_axes(rect5)
    
    pb_i = par_dict['pb_i']
    pb_o = par_dict['pb_o']

    wsrt_time=day_to_tasc(wsrt1400.mjd, base_mjd, tasc_o, pb_o)
    ax1.errorbar(wsrt_time, wsrt1400.phase*scl_mu, yerr=wsrt1400.unc*scl_mu, marker="o", linestyle='none', 
         mec=mec_w1400, markerfacecolor=face_w1400, color=err_w1400, markersize=my_ms, capsize=0, mew=my_mew)
    gbt_time=day_to_tasc(gbt.mjd, base_mjd, tasc_o, pb_o)
    ax1.errorbar(gbt_time, gbt.phase*scl_mu, yerr=gbt.unc*scl_mu, marker="o", linestyle='none', 
         mec=mec_gbt, markerfacecolor=face_gbt, color=err_gbt, markersize=my_ms, capsize=0, mew=my_mew)
    ao_time=day_to_tasc(ao1400.mjd, base_mjd, tasc_o, pb_o)
    ax1.errorbar(ao_time, ao1400.phase*scl_mu, yerr=ao1400.unc*scl_mu, marker="o", linestyle='none', 
         mec=mec_a1400, markerfacecolor=face_a1400, color=err_a1400, markersize=my_ms, capsize=0, mew=my_mew)

    mpl.rcParams['font.size'] = 7
    mpl.rcParams['font.family'] = 'sans-serif'
    mpl.rcParams['axes.linewidth'] = 1
    mpl.rcParams['ps.useafm'] = True
    mpl.rcParams['pdf.use14corefonts'] = True
    #mpl.rcParams['text.usetex'] = True
    mpl.rcParams['ytick.labelsize']=7
    mpl.rcParams['xtick.labelsize']=7
    mpl.rcParams['axes.labelsize']=7
    mpl.rcParams['axes.unicode_minus']=False
    mm_to_in = 0.0393701

    majorLocator = MultipleLocator(50)
    majorFormatter = FormatStrFormatter('%.0f')
    minorLocator = MultipleLocator(10)
    yearLocator = MultipleLocator(1)

    ax1.xaxis.set_major_formatter(majorFormatter)

    ax2.set_xlabel('Year')
    ax2.xaxis.set_major_formatter(majorFormatter)
    ax1.set_ylabel('Fit residuals \n ($\mu$s)')
    ax2.xaxis.set_major_locator(yearLocator)

    ax1.set_ylim(-14,14)
    ax1.set_xlim(-0.9,5.1)

    d_n, d_x = ax1.get_xlim()
    ax2.set_xlim(day_to_year(base_mjd), day_to_year(base_mjd+2010))
    ax1.xaxis.set_ticklabels([])

    my_red=np.array([123, 36, 28])/255.0

    tasc_time=day_to_tasc(new_times, base_mjd, tasc_o, pb_o)

    pb_i = par_dict['pb_i']
    pb_o = par_dict['pb_o']

    hex_bins=(int(x_bin/1.5),int(y_bin/1.5))

    fk_tasc_time=day_to_tasc(fk_times, base_mjd, tasc_o, pb_o)

    jm=ax3.hexbin(fk_tasc_time,(fk_times/pb_i)%1, fk_phase*scl_ns,hex_bins, linewidths=0.1,
                  vmin=-10000, vmax=10000, cmap='Greys', edgecolor=hex_edge, alpha=1., extent=(-1,6,0,1))
    im=ax3.hexbin(tasc_time,(new_times/pb_i)%1, new_phase*scl_mu, hex_bins, vmin=-1, vmax=1,
                  linewidths=0.1, extent=(-1,6,0,1), cmap=my_cmap)


    ax3.xaxis.set_ticklabels([])
    ax3.set_ylabel("Inner orbital phase \n (turns)")
    ax3.set_xlim(-0.9,5.1)

    norm1 = mpl.colors.Normalize(vmin=-1,vmax=1)
    cb1 = mpl.colorbar.ColorbarBase(ax5, cmap=my_cmap, norm=norm1, orientation='vertical')
    cb1.set_label('Fit residuals \n ($\mu$s)')
    cb1.ax.xaxis.set_ticks_position('none')


    fake_times=fk_times
    fk_tasc_time=day_to_tasc(fake_times, base_mjd, tasc_o, pb_o)

    hm=ax4.hexbin(fk_tasc_time,(fk_times/pb_i)%1, fk_only_delta*-1.1e-6*scl_ns,hex_bins, linewidths=0.1,
                  vmin=-130, vmax=130, cmap='Greys', edgecolor=hex_edge, alpha=1., extent=(-1,6,0,1))
    de=ax4.hexbin(tasc_time,(new_times/pb_i)%1, delta_only*scl_ns*deltaa,hex_bins,
                  cmap=my_cmap, linewidths=0.1, extent=(-1,6,0,1),  vmin=-100, vmax=100) 

    ax4.set_xlabel("Outer orbital phase \n (turns)")
    ax4.set_ylabel("Inner orbital phase \n (turns)")
    ax4.set_xlim(-0.9,5.1)

    norm2 = mpl.colors.Normalize(vmin=-80,vmax=80)
    cb1 = mpl.colorbar.ColorbarBase(ax6, cmap=my_cmap, norm=norm2, orientation='vertical')
    cb1.set_label('Signature of $\Delta$ \n (ns)')

    figsize=[width*mm_to_in,hight*mm_to_in]
    plt.gcf().set_size_inches(figsize[0], figsize[1])

    ex=0.07

    plt.figtext(x1-ex,y1+3.*h1+2.*gap,'a', family='sans-serif', weight='bold', size=8)
    plt.figtext(x1-ex,y1+2.*h1+gap,'b', family='sans-serif', weight='bold',  size=8)
    plt.figtext(x1-ex,y1+h1,'c', family='sans-serif', weight='bold',  size=8)

    print 'top corner:', y1+3.*h1+2.*gap
    plt.savefig('%s_%dmm.pdf'%(fig_name,width), format='pdf', dpi=300)
```
