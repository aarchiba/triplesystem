```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
from collections import namedtuple
from pylab import rcParams
rcParams['savefig.dpi'] = 144
import pickle
import math
from scipy.stats import norm
import matplotlib.mlab as mlab
from scipy import stats
from matplotlib.font_manager import FontProperties

import arrow_plot_3 as arw_pl
```

```python
data=pickle.load(open('systematics-package-lband-20170825.pickle', 'rb'))
```

###Our data

```python
phase=np.array(data['residuals'], dtype=np.float64)
base=np.array(data['base_mjd'], dtype=np.float64)
mjd=np.array(data['times'], dtype=np.float64)+base

unc=np.array(data['phase_uncerts'], dtype=np.float64)
par_dict=data['best_parameters']

pb_i = par_dict['pb_i']
pb_o = par_dict['pb_o']
phi_i=(2.0*math.pi)/pb_i
phi_o=(2.0*math.pi)/pb_o
tasc_i = par_dict["tasc_i"]
tasc_o = par_dict["tasc_o"]

plt.plot(mjd, phase, 'o', ms=2, ls='None', color='r')
print len(mjd)
plt.xlim(55900,58550)
im_width=15

plt.xlabel('MJD')
plt.ylabel('pulsar phase')

plt.gcf().set_size_inches(im_width,im_width*4./15.)

print np.amax(unc)
print np.amax(phase)

plt.savefig('test_res_all.pdf', bbox_inches="tight")
```

##Fake arrow insert

```python
#s = amplitude*cos(mjds)+np.random.randn(len(mjds))*phase_uncerts

def fake_arrow(mjd, ampl, unc_ampl, in_comp, out_comp):
    uncerts = unc_ampl*np.random.randn(len(mjd))
    ys = ampl*np.cos((in_comp*phi_i+out_comp*phi_o)*mjd+(in_comp*tasc_i+out_comp*tasc_o))+uncerts
    return ys
```

###Real data

```python
ampl=3e-5
unc_ampl=8e-7
k=-2
j=1


ys=fake_arrow(mjd,ampl,unc_ampl,k,j)
phase_2=phase+ys

value=arw_pl.arrow_plot(par_dict, mjd, phase_2, unc)
plt.title('s = ampl[=%1.0fe-5]*cos( [k*f_i+j*f_o] mjd)+random(len(mjd))*fake_unc[=8e-7]; k=%d, j=%d'%(ampl*1e5, k, j), fontsize=9)
plt.savefig('fake_unc8e-7_ampl%1.0fe-5_%d_%d.pdf'%(ampl*1e5, k, j))
plt.show()

ys=fake_arrow(mjd,ampl,unc,k,j)
phase_3=phase+ys

value=arw_pl.arrow_plot(par_dict, mjd, phase_3, unc)
plt.title('s = ampl[=%1.0fe-5]*cos( [k*f_i+j*f_o] mjd)+random(len(mjd))*real_unc; k=%d, j=%d'%(ampl*1e5, k, j), fontsize=9)
plt.savefig('real_unc_ampl%1.0fe-5_%d_%d.pdf'%(ampl*1e5, k, j))


print value
```

##Fake data

```python
ph_ampl=0.001
mjd_ampl=10

def fake_data(ph_ampl, mjd_ampl, mjd_start, mjd_fin, unc_ampl, lenght):
    fake_phase = ph_ampl*np.random.randn(lenght)
    fake_unc = unc_ampl*ph_ampl*np.random.randn(lenght)
    mjd_noise=mjd_ampl*np.random.randn(lenght)**2
    step=(mjd_fin-mjd_start)/lenght
    fake_mjd=np.arange(mjd_start, mjd_fin, step)

    fake_mjd=fake_mjd+mjd_noise+base
    fake_mjd=np.sort(fake_mjd)
    return fake_phase, fake_mjd, fake_unc 


scatter=3.5

fake_phase, fake_mjd, fake_unc = fake_data(0.001, 2.5, 0.0, 2000.0, 0.1, 20000.0)
fk_phase, fk_mjd, fk_unc = fake_data(0.0008, scatter, 0.0, 0.5, 0.1, 20)

for i in range(0,int(365/2.5)):
    f_phase, f_mjd, f_unc = fake_data(0.0008, scatter, 0.0+i*5, i*5+0.5, 0.4, 50)
    fk_phase=np.concatenate((fk_phase,f_phase), axis=0)
    fk_mjd=np.concatenate((fk_mjd,f_mjd), axis=0)
    fk_unc=np.concatenate((fk_unc,f_unc), axis=0)

for i in range(0,12*6):
    f_phase, f_mjd, f_unc = fake_data(0.0006, scatter, 0.0+i*30, i*30+0.5, 0.4, 50)
    fk_phase=np.concatenate((fk_phase,f_phase), axis=0)
    fk_mjd=np.concatenate((fk_mjd,f_mjd), axis=0)
    fk_unc=np.concatenate((fk_unc,f_unc), axis=0)

plt.xlim(55800,58200)    
    
print np.amax(fk_phase)
print np.amax(fk_unc)

plt.plot(fk_mjd, fk_phase, 'o', ms=2, ls='None')
plt.xlabel('MJD')
plt.ylabel('pulsar phase')
im_width=15
plt.gcf().set_size_inches(im_width,im_width*4./15.)

plt.savefig('test_fake_all.pdf', bbox_inches="tight")
```
