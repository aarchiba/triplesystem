```python
tel_out=namedtuple("tel_transform",
       ["phase", "mjd","unc", "index"])

def tel_transform(data,name,limiit):
    tel_res=np.array(data['residuals'][data['telescopes']==list(data['telescope_list']).index(name)], dtype=np.float64)
    tel_unc=np.array(data['phase_uncerts'][data['telescopes']==list(data['telescope_list']).index(name)], dtype=np.float64)
    tel_mjd=np.array(data['times'][data['telescopes']==list(data['telescope_list']).index(name)], dtype=np.float64)+old_base
    
    tel_index=(data['telescopes']==list(data['telescope_list']).index(name))
    
    tel_phase=tel_res[abs(tel_res/tel_unc) <limiit]
    tel_mjd=tel_mjd[abs(tel_res/tel_unc) <limiit]
    tel_unc=tel_unc[abs(tel_res/tel_unc) <limiit]
    
    return tel_out(phase=tel_phase, mjd=tel_mjd,unc=tel_unc, index=tel_index)

gbt=tel_transform(my_data,'GBT_1400',limiit)
ao1400=tel_transform(my_data,'AO_1400',limiit)
wsrt1400=tel_transform(my_data,'WSRT_1400',limiit)

#ao430=tel_transform(my_data,'AO_430',limiit)
#wsrt350=tel_transform(my_data,'WSRT_350',limiit)
```

```python
def lstsq_with_errors(A,b,uncerts=None):
    """Solve a linear least-squares problem and return uncertainties

    This function extends `scipy.linalg.lstsq` in several ways: first,
    it supports uncertainties on each row of the least-squares problem.
    Second, `scipy.linalg.lstsq` fails if the scales of the fit
    variables are very different. This function rescales them 
    internally to improve the condition number. Finally, this function
    returns an object containing information about the uncertainties
    on the fit values; the `uncerts` attribute gives individual
    uncertainties, the `corr` attribute is the matrix of correlations,
    and the `cov` matrix is the full covariance matrix.
    """
    if len(A.shape)!=2:
        raise ValueError
    if uncerts is None:
        Au = A
        bu = b
    else:
        Au = A/uncerts[:,None]
        bu = b/uncerts
    Ascales = np.sqrt(np.sum(Au**2,axis=0))
    #Ascales = np.ones(A.shape[1])
    if np.any(Ascales==0):
        raise ValueError("zero column (%s) in A" % np.where(Ascales==0))
    As = Au/Ascales[None,:]
    db = bu
    xs = None
    best_chi2 = np.inf
    best_x = None
    for i in range(5): # Slightly improve quality of fit
        dxs, res, rk, s = scipy.linalg.lstsq(As, db)
        if rk != A.shape[1]:
            raise ValueError("Condition number still too bad; "
                             "singular values are %s"
                             % s)
        if xs is None:
            xs = dxs
        else:
            xs += dxs
        db = bu - np.dot(As, xs)
        chi2 = np.sum(db**2)
        if chi2<best_chi2:
            best_chi2 = chi2
            best_x = xs
        #debug("Residual chi-squared: %s", np.sum(db**2))
    x = best_x/Ascales # FIXME: test for multiple b
    
    class Result:
        pass
    r = Result()
    r.x = x
    r.chi2 = best_chi2
    bias_corr = A.shape[0]/float(A.shape[0]-A.shape[1])
    r.reduced_chi2 = bias_corr*r.chi2/A.shape[0]
    Atas = np.dot(As.T, As)
    covs = scipy.linalg.pinv(Atas)
    r.cov = covs/Ascales[:,None]/Ascales[None,:]
    r.uncerts = np.sqrt(np.diag(r.cov))
    r.corr = r.cov/r.uncerts[:,None]/r.uncerts[None,:]
    return r
```

```python
#-----function-1 --- make a matrix
############################################
#input:
#ll - nhumber of inner orbit turns
#mjd - input array...

matrix_out=namedtuple("matrix",
       ["names","A","Adict"])


def matrix(ll, mjd):
    Adict = {}
    for i in range(ll):
        for j in range(-(ll-1),ll):
            if i==0 and j<0:
                continue
            Adict[i,j,'cos'] = np.cos((i*phi_o+j*phi_i)*mjd)
            if (i,j)!=(0,0):
                Adict[i,j,'sin'] = np.sin((i*phi_o+j*phi_i)*mjd)
    names = sorted(Adict.keys())
    A = np.array([Adict[n] for n in names]).T
    return matrix_out(names=names, A=A, Adict=Adict)
    
```

```python
#-----function-2 --- fit A using matrix function
###################################################

def linear_least_squares_cov(Acols, mjd, phase, unc):
    A = matrix(Acols, mjd)
    Adict=A.Adict
    names=A.names
    x, chi2, rk, s = scipy.linalg.lstsq(A.A/unc[:,np.newaxis], phase/unc)
    res = phase-np.dot(A.A,x)
    Ascaled = A.A/unc[:,np.newaxis]
    
    cov = np.linalg.pinv(np.dot(Ascaled.T,Ascaled))
    n=len(phase)
    cov_scaled = cov*chi2/n
    
    class Result(object):
        pass
    r = Result()
    r.names = names
    r.x = x
    r.chi2 = chi2
    r.rk = rk
    r.s = s
    r.res = rescov = np.linalg.pinv(np.dot(Ascaled.T,Ascaled))
    n=len(phase)
    r.A = A
    r.Adict = Adict
    r.cov = cov
    r.cov_scaled = cov_scaled
    
    return r
```

```python
#functions-3-and-4---draw-plots
###############################
plotter_out = namedtuple("plotter_out",
    ["X","Y","U","V","M","err_X","err_Y","err_ix"])

def plotter(Acols, mjd, phase, unc, color):
    ang = np.linspace(0,2*np.pi,100)
    circle = np.array([np.cos(ang),np.sin(ang)])
    fit = linear_least_squares_cov(Acols, mjd, phase, unc)
    result = dict(zip(fit.names,fit.x))
    X = []
    Y = []
    U = []
    V = []
    err_X = []
    err_Y = []
    err_ix = []
    for (i,j,f) in result.keys():
        Y.append(i)
        X.append(j)
        U.append(result[i,j,'cos'])
        err_ix.append((i,j))
        if (i,j)==(0,0):
            V.append(0)
            err_X.append(0)
            err_Y.append(0)
        else:
            V.append(result[i,j,'sin'])
            
            st=fit.names.index((i, j, 'cos'))
            fin=fit.names.index((i, j, 'sin'))+1
            err_M=fit.cov[st:fin,st:fin]
            assert err_M.shape==(2,2)
            L = scipy.linalg.cholesky(err_M)
            Lcircle = np.dot(L,circle)
            # print Lcircle[0][7]
            err_X.append(Lcircle[0])
            err_Y.append(Lcircle[1])
    
    M = np.hypot(V, U)
    return plotter_out(X=X, Y=Y, U=U, V=V, M=M, err_X=err_X, err_Y=err_Y, err_ix=err_ix)

err_grey=np.array([131, 145, 146])/255.0

def draw_plot(number, mjd, phase, unc, color):
    pl=plotter(number, mjd, phase, unc, color)
    ar_scale=np.amax(pl.M)
    for k in range(0,len(pl.err_X)):
        i,j = pl.err_ix[k]
        plt.plot(pl.err_X[k]/ar_scale+j,pl.err_Y[k]/ar_scale+i, color=color, lw=1, alpha=0.3)
    Q = plt.quiver(pl.X, pl.Y, pl.U, pl.V, pl.M, units='x', scale=ar_scale)
    plt.xlim(-number,number)
    plt.ylim(-1,number)
    plt.set_cmap('viridis')
    plt.xlabel('Inner orbit frequencies')
    plt.ylabel('Outer orbit frequencies')
    plt.gca().set_aspect('equal')

    plt.colorbar(label=r'$\mu$ s')
    return ar_scale


def draw_nocb_plot(number, mjd, phase, unc, color, my_width, note, scale=None):
    my_width=my_width
    mar_size=my_width*0.33
    lab_size=my_width*1.6
    tick_size=my_width*0.8
    font_size=my_width*2.
    
    font1 = FontProperties()
    font1.set_size('%d'%font_size)
    font1.set_family('serif')
    font1.set_style('normal')
    font1.set_weight('bold')
    plt.xlabel('Days',  fontproperties=font0)
    plt.tick_params('x', colors='k', size=tick_size)
    plt.tick_params('y', colors='k', size=tick_size)
    plt.rc('xtick', labelsize=lab_size) 
    plt.rc('ytick', labelsize=lab_size) 
    pl=plotter(number, mjd, phase, unc, color)
    
    if scale is None:
        ar_scale=np.amax(pl.M)
    else:
        ar_scale=scale
    for k in range(0,len(pl.err_X)):
        i,j = pl.err_ix[k]
        plt.plot(pl.err_X[k]/ar_scale+j,pl.err_Y[k]/ar_scale+i, color=color, lw=1, alpha=0.3)
    Q = plt.quiver(pl.X, pl.Y, pl.U, pl.V, units='x', scale=ar_scale, color=color)
    plt.xlim(-number,number)
    plt.ylim(-1,number)
    QQ = plt.quiver(-3, 0, 0.05, 0.0, units='x', scale=ar_scale, color=err_grey, alpha=0.8)
    #plt.set_cmap('Blues')
    plt.xlabel('Inner orbit frequencies', fontproperties=font1)
    plt.ylabel('Outer orbit frequencies', fontproperties=font1)
    plt.gca().set_aspect('equal')
    err_red=np.array([191, 54, 12])/255.0
    if note==True:
        font1.set_size('%d'%(font_size*0.8))
        plt.text(-3,0.1, r'50 ns'%(ar_scale), fontproperties=font1, color=err_grey)
    font1.set_size('%d'%font_size)
    #plt.colorbar()
    return ar_scale

err_red=np.array([191, 54, 12])/255.0
err_bl=np.array([21, 101, 192])/255.0
err_green=np.array([85, 139, 47])/255.0
err_violet=np.array([49, 27, 146])/255.0
```
