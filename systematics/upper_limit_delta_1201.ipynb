```python
#TOA error vs. post-fit resdiduals
scl=1e6*p_period

def plot_unc(res,unc,scl,cl_err,cl_face,cl_mec,name):
    plt.errorbar(unc*scl, res*scl, yerr=unc*scl, linestyle='none', markerfacecolor=cl_err, color=cl_err)
    plt.plot(unc*scl, res*scl, 'o', mec=cl_mec, markersize=2, linestyle='none', markerfacecolor=cl_face, label=name)

my_grey=np.array([213, 219, 219])/255.0
plot_unc(my_data['residuals'],my_data['phase_uncerts'],scl, my_grey,'lightgrey',my_grey,'with outliers')
plot_unc(wsrt1400.phase,wsrt1400.unc,scl,err_w1400,face_w1400,mec_w1400,'WSRT1400')
plot_unc(gbt.phase,gbt.unc,scl,err_gbt,face_gbt,mec_gbt,'GBT1500')
plot_unc(ao1400.phase,ao1400.unc,scl,err_a1400,face_a1400,mec_a1400,'AO1400')
#plot_unc(ao430.phase,ao430.unc,scl,err_a430,face_a430,mec_a430,'AO430')
#plot_unc(wsrt350.phase,wsrt350.unc,scl,err_w350,face_w350,mec_w350,'WSRT350')

plt.xscale('log')
plt.xlim(7e-2,3.5)
plt.ylim(-30,30)

lgd = plt.legend(numpoints=3, loc='upper left', bbox_to_anchor=(0.125, 1), fontsize=7)
plt.ylabel(r'residuals ($\mu$s)')
plt.xlabel(r'Unc. ($\mu$s)')

plt.gcf().set_size_inches(8,4.5)
```
