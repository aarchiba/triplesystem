```python
scl_mu=p_period*1e6#mus

def rms(y):
    return np.sqrt(np.sum(y**2.)/len(y))

def rms_w(y,w):
    new_w=1./w
    y_sum=0.
    w_sum=np.sum(new_w)
    newnew_w=new_w/w_sum
    for i in range(0,len(y)):
        pr=y[i]**2.*newnew_w[i]
        y_sum=y_sum+pr
    return np.sqrt(y_sum)

print 'std:'
print 'GBT  rms:',rms(gbt.phase*scl_mu),'wrms:',rms_w(gbt.phase*scl_mu,gbt.unc*scl_mu)
print 'AO   rms:',rms(ao1400.phase*scl_mu),'wrms:',rms_w(ao1400.phase*scl_mu,ao1400.unc*scl_mu)
print 'WSRT rms:',rms(wsrt1400.phase*scl_mu),'wrms:',rms_w(wsrt1400.phase*scl_mu,wsrt1400.unc*scl_mu)
print 'All rms:',rms(new_phase*scl_mu),'wrms:',rms_w(new_phase*scl_mu,new_unc*scl_mu)
print 'Uncertainty:'
print 'GBT  unc_rms:',rms(gbt.unc*scl_mu), 'unc_wrms:', rms_w(gbt.unc*scl_mu, gbt.unc*scl_mu)
print 'AO   unc_rms:',rms(ao1400.unc*scl_mu), 'unc_wrms:',rms_w(ao1400.unc*scl_mu,ao1400.unc*scl_mu)
print 'WSRT unc_rms:',rms(wsrt1400.unc*scl_mu), 'unc_wrms:',rms_w(wsrt1400.unc*scl_mu,wsrt1400.unc*scl_mu)
print 'All  unc_rms:',rms(new_unc*scl_mu), 'unc_wrms:',rms_w(new_unc*scl_mu,new_unc*scl_mu)
```
