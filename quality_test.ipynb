```python
def run(tol):
    tolrep = "tol = "+repr(tol)
    if tolrep in shelf:
        return
    times = shelf['times']
    initial_state = shelf['initial_state']
    O = quad_integrate.ODE(quad_integrate.KeplerRHS(),
                           initial_state, times[0],
                           0.01, atol=tol, rtol=tol)
    r = []
    for t in times:
        O.integrate_to(t)
        r.append(O.x)
    shelf[tolrep] = np.array(r)
    shelf["evals "+tolrep] = O.n_evaluations
```

```python
opi = kepler.pack_three_body(*orbital_parameters)
print orbital_parameters
print opi
positive_parameters = ['a_i','pb_i','a_o','pb_o']

def logify(op):
    l = list(op)
    for n in positive_parameters:
        i = kepler.three_body_parameters.index(n)
        l[i] = np.log(l[i])
    return tuple(l)
def unlogify(op):
    l = list(op)
    for n in positive_parameters:
        i = kepler.three_body_parameters.index(n)
        l[i] = np.exp(l[i])
    return tuple(l)

def shift_time(op, t):
    (a_i,pb_i,eps1_i,eps2_i,i_i,lan_i,q_i,tasc_i,
    a_o,pb_o,eps1_o,eps2_o,i_o,lan_o,tasc_o,
    x_cm,v_cm) = op
    return (a_i,pb_i,eps1_i,eps2_i,i_i,lan_i,q_i,tasc_i-t,
        a_o,pb_o,eps1_o,eps2_o,i_o,lan_o,tasc_o-t,
        x_cm,v_cm)

def fitfunc(op):
    #op = unlogify(op)
    opu = kepler.unpack_three_body(op)
    k_zs = [kepler.kepler_three_body(*shift_time(opu,t))[0][2] for t in times]
    k_zs = np.array(k_zs)
    return (k_zs - best_z).astype(float)

def dfunc(op):
    #op = unlogify(op)
    opu = kepler.unpack_three_body(op)
    k_z_d = [kepler.kepler_three_body(*shift_time(opu,t))[1][2,:] for t in times]
    k_z_d = np.array(k_z_d)
    return k_z_d.astype(float)
    
```

Measurability
-------------

We should be able to judge the measurability of various parameters by looking at the Jacobian. Parameters or parameter combinations that appear with small condition numbers are going to be a challenge. Of course, if we're getting a Jacobian from the Keplerian model, we'll only get to see the things measurable in a pure Keplerian model, that is, without gravitational interaction between inner and outer orbits.

```python
asini_i = a_i*np.sin(i_i)
acosi_i = a_i*np.cos(i_i)
asini_o = a_o*np.sin(i_o)
acosi_o = a_o*np.cos(i_o)
delta_lan = lan_o-lan_i

orbital_parameters_m = (asini_i, pb_i, eps1_i, eps2_i, tasc_i, 
        acosi_i, q_i,
        asini_o, pb_o, eps1_o, eps2_o, tasc_o, 
        acosi_o, delta_lan,
        lan_i, x_cm, v_cm)
```
