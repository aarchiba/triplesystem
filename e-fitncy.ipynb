```python
tel_list = []
tels = []
for l in open(timfile).readlines():
    if len(l)<10:
        continue
    ic = l.split()[0]
    if ic in ["1","3","i","f","@"]:
        if ic not in tel_list:
            tel_list.append(ic)
        tels.append(tel_list.index(ic))
if len(tels)!=len(r.bary_TOA):
    raise ValueError("Unable to assign telescope codes correctly")
tels = np.array(tels)
```

```python
i = F2.tel_list.index('NCY1400')
c = F2.tels==i
resids = r[c]
```

```python
toas = []
i = 0
for l in open("toas/ncy/0337.NRT-t2-201404507").readlines():
    s = l.split()
    if len(s)<3:
        continue
    flags = dict(zip([ss[1:] for ss in s[5::2]],s[6::2]))
    flags['freq'] = float(s[1])
    flags['mjd'] = float(s[2])
    flags['err'] = float(s[3])
    flags['filename'] = s[0]
    f = s[0].split(".")
    flags['calib'] = f[1]
    flags['proc'] = f[2]
    flags['band'] = "S" if flags['freq']>2000 else "L"
    flags['resid'] = resids[i]
    i += 1
    toas.append(flags)
    
print len(toas), np.sum(F2.tels==F2.tel_list.index("NCY1400"))
toas[:3]
```

```python
bws = set(t['bw'] for t in toas)
calibs = set(t['calib'] for t in toas)
bands = set(t['band'] for t in toas)

bcbs = set((t['band'],t['calib'],t['bw']) for t in toas)
bcbs
```

```python
for band, calib, bw in sorted(bcbs):
    x, y = zip(*[(t['mjd'],t['resid']) for t in toas
                 if t['band']==band and t['calib']==calib and t['bw']==bw])
    plt.plot(x,y,".", markersize=2, 
             label="/".join([band,calib,bw]))
plt.legend(loc='best')
plt.xlabel("MJD")
plt.ylabel("Residual (turns)")
plt.savefig("ncy-problem.pdf")
```
