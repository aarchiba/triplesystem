Timescales and delay correction for the triple system
=====================================================

Let $E_n$ be the event of the emission of the nth pulse. Let $t_p$ be the pulsar proper time, so that 
$$
F_0\cdot t_p(E_n)+F_1\cdot (t_p(E_n))^2/2 = n. 
$$
Let $t_d$ be the dynamical timescale of the system, that is, the clock by which the Newtonian orbital mechanics evolve. Then 
$$
t_d(E_n) = \int_0^{t_p(E_n)} \gamma^{-1}(t_p) dt_p \approx t_p(E_n) - \int_0^{t_p(E_n)} (\gamma-1) dt_p.
$$
Finally, $t_b$ is the time at which the signal would reach the solar system barycenter (ignoring all propagation and viewing-angle effects). So
$$
t_b(E_n) = t_d(E_n)+z(t_d(E_n))/c.
$$


When we observe the TOA corresponding to pulse $n$, we are (after correcting for various solar system and propagation effects) measuring $t_b(E_n)+i(E_n)$, where $i$ is an intrumental term; for our purposes we will assume that $i(E_n)$ simply returns a delay whose value is determined by the telescope/backend/frequency combination. 

What we have as input to the fitting is a set of delays $d_n$ approximately equal to $(t_b(E_n)+i(E_n))-t_p(E_n)$ and a set of corresponding "observation times" $t_b(E_n)+i(E_n)$ which have been modified by about 0.25 microseconds by roundoff error. I say the delays are "approximate" because in fact we don't know the correct F0 and F1.

If we have an error in F0 and F1, then that error will affect our computation of $t_p(E_n)$.

We need to compute $t_d(E_n)$ so that we can compute $z(t_d(E_n))$ by integration.
