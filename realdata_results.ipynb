```python
def plot_residuals(res,
    xbins=20,
    ybins=24,
    shift = 0.):
    r2 = threebody.remove_trend(res, mjds, tel_list, tels)
    print np.sqrt(np.mean(r2**2))
    plt.cool()
    ns = kepler.three_body_parameters_measurable
    pb_i = orbital_parameters_fit_g2[ns.index('pb_i')]
    pb_o = orbital_parameters_fit_g2[ns.index('pb_o')]
    tasc_o = orbital_parameters_fit_g2[ns.index('tasc_o')]
    tasc_i = orbital_parameters_fit_g2[ns.index('tasc_i')]
    phases_i = ((mjds-tasc_i)/pb_i-shift)%1+shift
    wt, xedges, yedges = np.histogram2d(phases_i, mjds,
        [xbins,ybins], [[0+shift,1+shift],[np.amin(mjds),np.amax(mjds)]])
    res2d, xedges, yedges = np.histogram2d(phases_i, mjds,
        [xbins,ybins], [[0+shift,1+shift],[np.amin(mjds),np.amax(mjds)]],
        weights=np.asarray(r2*1e6,float))
    r = ma.array(res2d/wt)
    r[wt==0] = ma.masked
    plt.pcolormesh(xedges, yedges/365.2425, r.T) # FIXME: double-check orientation
    plt.ylim(np.amin(mjds)/365.2425,
        np.amax(mjds)/365.2425)
    plt.ylabel("time (y)")
    plt.xlabel("inner orbital phase")
    plt.xlim(shift, 1+shift)
    plt.colorbar(orientation="horizontal")

    plt.twinx()
    def mtoo(m):
        return (m-tasc_o)/pb_o+1
    plt.ylim(mtoo(np.amin(mjds)),
        mtoo(np.amax(mjds)))
    plt.ylabel("outer orbital phase")
```

What about non-binary parameters? If the pulsar period is wrong, that can be absorbed into v_cm[2]. If the pulsar spin-down is wrong, that will remain as a quadratic term. If the position is wrong, that gives us a one-year sinusoid. If the parallax is wrong, that gives us a half-year sinusoid. If the proper motion is wrong, that gives us a one-year sinusoid increasing with t. 

It's also worth noting that since we're down at the microsecond level, the accumulated integration error could be getting important. It may be worth rerunning this with different tolerances. We might also be being bitten by uncertainties in tke Keplerian model used for the initial conditions. Maybe worth improving this to long double and checking its limitations.

Relativistic effects: special relativistic time dilation, general relativistic time dilation. Together these give what is called the Einstein delay in pulsar timing.

```python
c = 86400.
v2 = np.sqrt(np.sum(fit_solution[:,3:6]**2, axis=-1))
special_slowing = (1-v2/c**2)**(-0.5)
def general_slowing(body1, body2):
    r = np.sqrt(np.sum((body2[...,:3]-body1[...,:3])**2,axis=-1))
    m2 = body2[...,6]
    return (1-2*kepler.G*m2/(r*c**2))**(-0.5)
```
