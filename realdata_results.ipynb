```python
delay_list = []
with open("0337_delays_2.txt") as f:
    for l in f.readlines():
        if l.startswith("#"):
            continue
        mjd, delay, tel = l.split()
        delay_list.append((float(mjd),float(delay),tel))
mjds = np.array([m for (m,d,t) in delay_list])
delays = np.array([d for (m,d,t) in delay_list])
tel_list = list(sorted(set([t for (m,d,t) in delay_list])))
tels = np.array([tel_list.index(t) for (m,d,t) in delay_list])
mjds -= delays/86400.
ix = np.argsort(mjds)
mjds = mjds[ix]
delays = delays[ix]
tels = tels[ix]
```

```python
initial_state = kepler.kepler_three_body_measurable(*(list(orbital_parameters_fit_g2)+[0]))[0]

tol = 1e-16
O = quad_integrate.ODE(quad_integrate.KeplerRHS(),
                           initial_state, 0,
                           rtol = tol, atol = tol)
fit_solution = []
for t in mjds:
    O.integrate_to(t)
    fit_solution.append(O.x)
fit_solution = np.array(fit_solution)
```

```python
year_length = 365.2425
def remove_trend(vec,
    P=True, Pdot=True, jumps=True,
    position=True, proper_motion=False, parallax=False):
    
    non_orbital_basis = [np.ones_like(mjds)]
    if P:
        non_orbital_basis.append(mjds)
    if Pdot:
        non_orbital_basis.append(mjds**2/2.)
    if jumps:
        non_orbital_basis.append(np.arange(1,len(tel_list))[:,None]==tels[None,:])
    if position:
        non_orbital_basis.extend([np.cos(2*np.pi*mjds/year_length),
                                  np.sin(2*np.pi*mjds/year_length)])
    if proper_motion:
        non_orbital_basis.extend([mjds*np.cos(2*np.pi*mjds/year_length),
                                  mjds*np.sin(2*np.pi*mjds/year_length)])
    if parallax:
        non_orbital_basis.extend([np.cos(4*np.pi*mjds/year_length),
                                  np.sin(4*np.pi*mjds/year_length)])
    non_orbital_basis = np.vstack(non_orbital_basis).T
    x, res, rk, s = scipy.linalg.lstsq(non_orbital_basis, vec)
    return vec-np.dot(non_orbital_basis,x)
```

```python
def shapiro_delay(s_src, s_m):
    c = 86400. # lt-s per day
    dx = s_m[...,:3]-s_src[...,:3]
    ldx = np.sqrt(np.sum(dx**2,axis=-1))
    return -86400*2*kepler.G*s_m[...,6]*c**(-3)*np.log(1-dx[...,2]/ldx) # in s
```

What about non-binary parameters? If the pulsar period is wrong, that can be absorbed into v_cm[2]. If the pulsar spin-down is wrong, that will remain as a quadratic term. If the position is wrong, that gives us a one-year sinusoid. If the parallax is wrong, that gives us a half-year sinusoid. If the proper motion is wrong, that gives us a one-year sinusoid increasing with t. 

It's also worth noting that since we're down at the microsecond level, the accumulated integration error could be getting important. It may be worth rerunning this with different tolerances. We might also be being bitten by uncertainties in tke Keplerian model used for the initial conditions. Maybe worth improving this to long double and checking its limitations.
